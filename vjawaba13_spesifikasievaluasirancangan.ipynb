# Spesifikasi Evaluasi Rancangan
Evaluasi rancangan dalam penelitian ini dilakukan secara berlapis dan komprehensif menggunakan kerangka kerja PICOC-ASTF. Kerangka ini memungkinkan validasi pada setiap lapisan, dari konsep fundamental hingga aplikasi di dunia nyata.

* Population: Siapa yang diteliti (mahasiswa, dosen).
* Intervention: Apa intervensi yang diberikan (misalnya, CKMS-SE).
* Comparison: Apa pembandingnya (misalnya, kelas konvensional).
* Outcome: Apa hasil yang diukur (misalnya, skor VCC, motivasi).
* Context: Di mana penelitian dilakukan (misalnya, mata kuliah Probabilitas dan Statistika di ITB).

Evaluasi ini diterapkan pada empat lapisan arsitektur ASTF:

* Lapisan Fundamental (F - Teoretis):
    * Fokus: Mengevaluasi koherensi dan validitas model konseptual (integrasi VCC-VOE, TISE).
    * Metode: Tinjauan Literatur Sistematis (SLR) dan Validasi  konsep oleh Panel Ahli (Expert Judgment) dari 6 ahli dengan target rating rata-rata ≥4.0/5.0
* Lapisan Teknologi (T - Komponen AI):
    * Fokus: Menguji kinerja komponen teknologi, terutama PUDAL Engine.
    * Metode: Pengujian teknis untuk memverifikasi akurasi algoritma AI dalam melakukan personalisasi (target akurasi profiling ≥85%) 
* Lapisan Sistem (S - Ketergunaan Prototipe):
    * Fokus: Mengukur kemudahan penggunaan dan kualitas kolaborasi dari prototipe CKMS-SE.
    * Metode: Uji Ketergunaan (Usability Testing) dengan n=20 pengguna (mahasiswa dan dosen) menggunakan System Usability Scale (SUS) dengan target skor ≥70 
* Lapisan Aplikasi (A - Dampak Empiris):
    * Fokus: Memvalidasi efektivitas sistem secara keseluruhan dalam mencapai tujuan pembelajaran.
    * Metode: Quasi-Experiment (n=135) membandingkan Kelas Eksperimen vs Kontrol untuk mengukur Learning Gain dan Intrinsic Motivation (menggunakan kuesioner IMI), identitas profesional, dan VCC dibandingkan dengan kelompok kontrol
