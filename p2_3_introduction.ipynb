# PENDAHULUAN

Profesi rekayasa menghadapi transformasi yang belum pernah terjadi sebelumnya. Meskipun mahasiswa berhasil menguasai konten teknis sebagian besar tidak siap untuk tantangan rekayasa yaitu kemampuan merancang solusi di tengah konteks sosial yang kompleks, membuat keputusan etis dalam ketidakpastian, dan berkolaborasi secara efektif lintas batas disiplin. Pada saat yang sama, Kecerdasan Buatan mengancam untuk mengotomatisasi kompetensi teknis yang ditekankan dalam pendidikan rekayasa. Sebuah laporan World Economic Forum 2024 mencatat bahwa 68 % eksekutif melaporkan kesenjangan keterampilan AI pada tingkat moderate-to-ekstrem. Hasil Organisation for Economic Co-operation and Development (OECD) Programme for International Student Assessment (PISA) menunjukkan penurunan catatan dalam kinerja siswa pada keterampilan matematika, membaca, dan sains. Selain itu, pasar tenaga kerja masa depan menuntut keterampilan kognitif (seperti berpikir analitis dan kreatif), keterampilan teknologi (seperti AI dan literasi Big Data), serta keterampilan interpersonal (seperti kepemimpinan dan empati) (REFERENSI : Forum, 2024).

Konvergensi ini mengungkapkan kesenjangan pedagogis mendasar: pembelajaran tradisional berfokus pada pencapaian kompetensi dalam batas materi, sumber daya, dan durasi waktu yang terbatas tanpa merumuskan dampak yang diharapkan dari kompetensi tersebut. Munculnya AI dan akses pengetahuan yang murah dan merata secara fundamental telah menggeser kebutuhan pendidikan dari sekadar replikasi pengetahuan menjadi pengelolaan pengetahuan yang berfokus pada pemecahan masalah kompleks yang memiliki nilai organisasi atau sosial yang signifikan. Penelitian ini mengusulkan platform yang mengintegrasikan Pembelajaran Kolaboratif dan Pembelajaran Terpersonalisasi yang disinergikan oleh AI dalam konsep Edukasi Berorientasi Nilai (VOE) untuk mendorong Penciptaan Nilai Bersama (VCC) yang sistematis.

Kerangka Pembelajaran VALORIZE muncul sebagai respons yang kuat, mengusulkan pergeseran radikal dalam tujuan pendidikan dari penguasaan konten menjadi “pembentukan sosok, karakter, dan pola pikir profesi” (menjadi sosok) yang secara langsung menerapkan prinsip TISE Homocordium. Homocordium mewakili nilai, etika, kreativitas, dan tujuan manusia. Dengan berfokus pada menjadi sosok, kerangka ini menumbuhkan penilaian ahli dan penalaran etis sebagai proses yang secara inheren manusiawi dan tahan terhadap otomatisasi memberikan proposisi nilai yang tahan lama untuk pendidikan tinggi di Era AI.

Paper ini mengusulkan Collaborative Knowledge Management System - Smart Engineering (CKMS-SE), sebuah Artefak Edukasi Cerdas yang dibangun sebagai validasi empiris dari sintesis TISE-VALORIZE. Pertanyaan penelitian ini adalah: Apakah sistem manajemen pengetahuan kolaboratif yang dirancang melalui prinsip TISE-VALORIZE, yang menggabungkan pemodelan kognitif formal dan alur kerja profesional, meningkatkan pembentukan identitas profesional dan penciptaan nilai yang terukur dalam pendidikan rekayasa?

Kontribusi utama penelitian ini adalah: (1) Mengoperasionalkan TISE sebagai Filosofi yang Dapat Dieksekusi. Kami menyajikan CKMS-SE sebagai instansiasi empiris pertama TISE-VALORIZE, yang secara formal menghubungkan prinsip abstrak Homocordium dan Triune-Intelligence ke praktik pedagogis yang terukur melalui siklus Perceive, Understand, Decision-making, Act, Learning (PUDAL) dan nilai Product, Service, Knowledge, Value, Environment (PSKVE) yang komprehensif. (2) Meningkatkan Fidelitas Kognitif melalui Taksonomi Multi-Graf: Kami meningkatkan mesin kognitif menggunakan Taksonomi Multi-Graf yang bergerak dari peta pemecahan masalah metaforis ke struktur graf formal yang ketat yang secara eksplisit memodelkan dan menilai proses kognitif multi-tahap dari insinyur ahli (peningkatan VALORIZE 2.0). (3) Menyematkan Alur Kerja Profesional dan Nilai Holistik: Kami merekayasa ulang Knowledge Marketplace menjadi Portofolio Nilai PSKVE Multi-Aset, yang menghargai produktivitas, keandalan, dan dampak sosial, serta mengintegrasikan simulasi profesi dengan alur kerja Agile/Scrum dan Git/GitHub untuk menciptakan catatan praktik profesional kolaboratif yang otentik dan dapat diverifikasi (peningkatan VALORIZE 2.0).

# INTRODUCTION

The engineering profession is facing unprecedented transformation. Although students successfully master technical content, they are largely unprepared for the engineering challenge of designing solutions within complex social contexts, making ethical decisions in uncertainty, and collaborating effectively across disciplinary boundaries. At the same time, Artificial Intelligence threatens to automate the technical competencies emphasised in engineering education. A 2024 World Economic Forum report notes that 68% of executives report AI skills gaps at a moderate-to-extreme level. Additionally, the future labour market demands cognitive skills (such as analytical and creative thinking), technological skills (such as AI and Big Data literacy), and interpersonal skills (such as leadership and empathy) (REFERENCE: Forum, 2024).

This convergence reveals a fundamental pedagogical gap: traditional learning focusses on achieving competence within limited material, resource, and time constraints without articulating the expected impact of that competence. The emergence of AI and affordable, equitable access to knowledge have fundamentally shifted the need for education from mere knowledge replication to knowledge management focused on solving complex problems with significant organisational or social value. This research proposes a platform that integrates Collaborative Learning and Personalised Learning, synergised by AI within the Value-Oriented Education (VOE) concept, to systematically promote Value Co-Creation (VCC).

The VALORIZE Learning Framework emerges as a powerful response, proposing a radical shift in educational goals from content mastery to "shaping the figure, character, and mindset of the profession" (becoming a figure), directly applying the principles of TISE Homocordium. Homocordium represents human values, ethics, creativity, and purpose. By focussing on becoming a figure, this framework cultivates expert judgement and ethical reasoning as inherently human processes resistant to automation, providing a lasting value proposition for higher education in the AI Era.

This paper proposes the Collaborative Knowledge Management System - Smart Engineering (CKMS-SE), an Intelligent Educational Artefact built as an empirical validation of the TISE-VALORIZE synthesis. The research question is: Does a collaborative knowledge management system designed through the TISE-VALORIZE principles, which combines formal cognitive modelling and professional workflows, enhance professional identity formation and measurable value creation in engineering education?

The main contributions of this research are: (1) Operationalising TISE as an Executable Philosophy. We present CKMS-SE as the first empirical instantiation of TISE-VALORIZE, formally linking the abstract principles of Homocordium and Triune-Intelligence to measurable pedagogical practices through the Perceive, Understand, Decision-making, Act, Learning (PUDAL) cycle and comprehensive Product, Service, Knowledge, Value, Environment (PSKVE) values. (2) Enhancing Cognitive Fidelity through Multi-Graph Taxonomies: We enhanced the cognitive engine using Multi-Graph Taxonomies that move from metaphorical problem-solving maps to strict formal graph structures that explicitly model and assess the multi-stage cognitive processes of expert engineers (VALORIZE 2.0 enhancement). (3) Embedding Professional Workflows and Holistic Value: We re-engineered the Knowledge Marketplace into the PSKVE Multi-Asset Value Portfolio, which values productivity, reliability, and social impact, and integrates professional simulations with Agile/Scrum and Git/GitHub workflows to create authentic and verifiable collaborative records of professional practice (VALORIZE 2.0 enhancement).
