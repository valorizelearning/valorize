# Hasil Penelitian

## Arsitektur Sistem CKMS-SE

Arsitektur ini mengimplementasikan siklus kognitif berkelanjutan melalui dua mesin khusus yang terintegrasi dalam paradigma rekayasa cerdas.
1, Mesin PUDAL (Perceive-Understand-Decision-Act-Learning-Evaluate). Mesin ini mengimplementasikan siklus perbaikan kognitif berkelanjutan. Mesin ini bertindak sebagai “mesin mikro-evolusioner” yang memaksa perbaikan dan adaptasi berkelanjutan.
2. Mesin PSKVE (Product-Service-Knowledge-Value-Environmental). Mesin ini mengelola dan mengoptimalkan Penciptaan Nilai Bersama (VCC) dengan memastikan bahwa nilai yang diciptakan melampaui pembelajaran individu ke pembangunan aset komunitas yang dapat diverifikasi. Sistem ini terdiri dari tiga elemen utama yang saling terhubung: Core Engine sebagai mekanisme sentral, AI-based PUDAL Engine untuk Pembelajaran Terpersonalisasi, dan PSKVE Engine untuk manajemen Penciptaan Nilai Bersama (VCC).

# Research Findings

## CKMS-SE System Architecture

This architecture implements a continuous cognitive cycle through two specialised engines integrated within the intelligent engineering paradigm.
1. PUDAL Engine (Perceive-Understand-Decision-Act-Learning-Evaluate). This engine implements a continuous cognitive improvement cycle. This machine acts as a "micro-evolutionary engine" that forces continuous improvement and adaptation.
2. The PSKVE (Product-Service-Knowledge-Value-Environmental) machine. This machine manages and optimises Shared Value Creation (SVC) by ensuring that the value created goes beyond individual learning to the development of verifiable community assets. This system consists of three interconnected main elements: the Core Engine as the central mechanism, the AI-based PUDAL Engine for Personalised Learning, and the PSKVE Engine for managing Co-Creation of Value (VCC).
