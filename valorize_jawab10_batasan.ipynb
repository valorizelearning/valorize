### Batasan Spesifik Penelitian CKM-SE

Penelitian ini secara sadar menetapkan batasan-batasan yang jelas untuk memastikan validitas hasil pada tahapannya saat ini. Batasan-batasan tersebut tidak mengurangi nilai temuan, melainkan mendefinisikan ruang lingkup di mana temuan tersebut paling valid.

---

#### 1. Batasan Konteks dan Generalisasi

*   **Institusi Tunggal**: Studi empiris utama dan uji coba terkontrol dilaksanakan secara spesifik di **Institut Teknologi Bandung (ITB)** [ @76, @30, @93 ]. Hal ini berarti temuan belum tentu dapat digeneralisasikan secara langsung ke institusi lain dengan budaya akademik, sumber daya, atau profil mahasiswa yang berbeda.
*   **Disiplin Ilmu Spesifik**: Intervensi difokuskan pada mahasiswa sarjana teknik dan diuji pada mata kuliah inti seperti **Probabilitas & Statistika** dan **Rekayasa Sistem** [ @76 ]. Efektivitas kerangka kerja ini pada disiplin ilmu non-teknik (misalnya, ilmu sosial atau humaniora) belum diuji dalam penelitian ini.
*   **Skalabilitas Belum Teruji Penuh**: Penelitian ini mengakui bahwa sebagian besar riset inovasi pembelajaran, termasuk yang ini, masih merupakan uji coba di satu institusi dan satu mata kuliah [ @25 ]. Masalah skalabilitas dan keberlanjutan multi-tahun merupakan agenda untuk penelitian di fase berikutnya [ @69, @122 ].

#### 2. Batasan Cakupan Teknologi

*   **Fokus pada Mesin Inti**: Implementasi teknologi secara spesifik berfokus pada dua komponen utama: **PUDAL Engine** untuk personalisasi pembelajaran dan **PSKVE Engine** untuk pengukuran nilai [ @1, @13 ]. Fitur-fitur lain atau integrasi yang lebih dalam dengan sistem manajemen pembelajaran (LMS) universitas yang ada belum menjadi fokus utama pada tahap ini.

#### 3. Batasan Durasi dan Tahapan Penelitian

*   **Tahapan DSRM Terbatas**: Penelitian ini secara eksplisit dibatasi pada tiga fase pertama dari *Design Science Research Methodology* (DSRM), yang mencakup periode **17 bulan** dari Agustus 2025 hingga Desember 2026 [ @13 ].
*   **Pengukuran Jangka Pendek-Menengah**: Uji coba quasi-eksperimental berlangsung selama satu semester (12 minggu) [ @76 ]. Oleh karena itu, penelitian ini mengukur dampak jangka pendek hingga menengah pada variabel seperti motivasi intrinsik dan transfer pembelajaran. Dampak jangka panjang, seperti pengaruhnya terhadap kinerja karier setelah lulus, berada di luar cakupan studi ini.

#### 4. Batasan Terkait Karakteristik Peserta

*   **Tidak Efektif untuk Semua Peserta**: Penelitian ini secara transparan melaporkan bahwa kerangka kerja ini tidak memberikan hasil maksimal bagi semua peserta secara seragam. Ditemukan bahwa **sekitar 13% mahasiswa** menunjukkan peningkatan yang minimal [ @81 ].
*   **Penyebab yang Teridentifikasi**: Kegagalan ini bukan disebabkan oleh kelemahan kerangka kerja secara umum, melainkan oleh faktor-faktor spesifik pada peserta, yaitu:
    1.  **Literasi Digital Awal yang Rendah**: Beberapa mahasiswa mengalami kesulitan beradaptasi dengan alat-alat standar industri seperti **Git/GitHub** [ @81 ].
    2.  **Ketidaknyamanan dengan Evaluasi Sejawat**: Adanya resistensi atau ketidaknyamanan terhadap proses *peer review* menghambat keterlibatan penuh dalam siklus kolaborasi [ @81 ].

Batasan ini mengidentifikasi sebuah *boundary condition*, yaitu kondisi di mana implementasi CKM-SE memerlukan dukungan tambahan seperti pelatihan teknis wajib dan perancah (*scaffolding*) yang berbeda untuk memastikan semua mahasiswa dapat berpartisipasi secara efektif [ @81 ].
