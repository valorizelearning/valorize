# Batasan Penelitian
Untuk menjaga fokus dan kelayakan (doable), batasan penelitian ini adalah:

* Konteks Akademik: Penelitian terbatas pada mahasiswa program sarjana di Sekolah Teknik Elektro dan Informatika (STEI) Institut Teknologi Bandung (ITB) dengan fokus pada mata kuliah seperti Probabilitas dan Statistika. Validasi efektivitas dilakukan dalam durasi 1 semester akademik.
* Durasi: Penelitian dibatasi pada tiga fase DSRM (SK I, II, dan III) yang dilaksanakan selama periode 17 bulan, dari Agustus 2025 hingga Desember 2026
* Fokus Teknologi: Implementasi teknologi difokuskan pada pengembangan dua komponen utama: PUDAL Engine untuk personalisasi dan PSKVE Engine untuk pengukuran nilai. Aspek teknologi lain tidak menjadi fokus utama
* Metrik Kinerja: Keberhasilan diukur menggunakan target kuantitatif yang spesifik, seperti System Usability Scale (SUS) ≥70, reliabilitas antar-penilai kappa Cohen ≥0,75, dan ukuran efek (Cohen&#39;s d) ≥0,80 untuk hasil belajar utama

Batasan ini memastikan bahwa klaim penelitian dapat diuji secara empiris dalam lingkup yang terkendali, meskipun bertujuan untuk menghasilkan kerangka kerja yang dapat digeneralisasi.
