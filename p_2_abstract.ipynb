# Abstract (English UK) Grammarly 0% tapi kalimat lebih dari 160 kata

**Abstract**

The engineering profession is facing an existential crisis due to the automation of Artificial Intelligence (AI). Automation is eroding routine technical skills and exposing pedagogical gaps in fostering unique human competencies. In response, this research presents the Collaborative Knowledge Management System-Smart Engineering (CKMS-SE) framework. CKMS-SE serves as a validated operational blueprint for the Triune-Intelligence Smart Engineering (TISE, which integrates human, artificial, and natural intelligence, and VALORISE (a philosophy focused on value-oriented learning). Through its dual-engine architecture, meaning it combines two core processing units, likely human and AI, the system synergises human, artificial, and natural intelligence. This enables personalised learning and the creation of shared value. A quasi-experimental study (N = 143) using Design Science Research Methodology (DSRM, a framework for solving complex problems and designing innovative artefacts) revealed a substantial impact on psychological domains, especially intrinsic motivation (inner drive to learn) and professional identity (sense of self in the profession). The large-scale experimental group (K2, N = 56) showed a 12.6% advantage in cognitive performance (learning outcomes) over the baseline control group (Mean 24.93 vs 22.14). This confirms the intervention's effectiveness in a high-variability population (ANOVA F = 13.34; p < 0.001). The findings demonstrate the TISE-VALORIZE framework's ability to shift education's focus from content mastery to professional development in essential humanistic capabilities in the AI Era.

Keywords: Artificial intelligence, CKMS-SE, engineering education, professional identity, TISE, VALORIZE.
