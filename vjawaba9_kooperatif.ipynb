# Perbedaan Collaborative vs. Cooperative Learning

* *Cooperative Learning*: Berfokus pada pembagian tugas dan akuntabilitas individu. Tujuannya adalah penyelesaian tugas efisien. "Saya mengerjakan bagian A, kamu B, lalu kita gabung."

* *Collaborative Learning* (Pendekatan Riset Ini): Berfokus pada konstruksi makna bersama (mutual engagement). Tujuannya adalah penciptaan pengetahuan baru yang tidak bisa dicapai sendirian. "Mari kita diskusikan masalah ini bersama hingga kita menemukan solusi yang kita berdua pahami." Riset ini menggunakan Kolaboratif karena fokus pada Value Co-Creation.

Intinya, CKM-SE melampaui sekadar kerja sama untuk menyelesaikan tugas. CKM-SE menciptakan sebuah ekosistem di mana pengetahuan secara aktif diciptakan, dibagikan, dan dinilai sebagai aset komunitas.

## Perbandingan: Pembelajaran Kooperatif vs. Pembelajaran Kolaboratif (CKMS-SE)

Tabel berikut menyajikan perbandingan mendalam antara dua pendekatan pedagogis: Pembelajaran Kooperatif (CL) sebagai model yang lebih tradisional dan Pembelajaran Kolaboratif sebagaimana diimplementasikan dalam kerangka kerja CKMS-SE, yang mengintegrasikan teknologi modern dan prinsip-prinsip ko-kreasi.

| Dimensi | Pembelajaran Kooperatif (Cooperative Learning) | Pembelajaran Kolaboratif (dalam Kerangka CKMS-SE) |
| :--- | :--- | :--- |
| **Tujuan (Objective)** | Berfokus pada penyelesaian tugas dengan **akuntabilitas individu** yang kuat dan sering kali pembagian imbalan. Tujuannya adalah agar setiap anggota kelompok dapat menunjukkan penguasaan materi secara mandiri. | Bertujuan untuk **konstruksi pengetahuan** dan **penciptaan nilai bersama** (*Value Co-Creation*/VCC) secara kolektif. Proses ini bersifat timbal balik dan menggabungkan sumber daya institusional serta pelajar untuk menciptakan nilai bersama (Robayo-Pinzon et al., 2024; Prahalad & Ramaswamy, 2004). |
| **Struktur (Structure)** | Biasanya sangat terstruktur, melibatkan **pembagian tugas** (*division of labor*) dan peran yang ditentukan secara eksplisit. Setiap anggota bertanggung jawab atas bagian spesifik dari tugas keseluruhan. | Lebih cair dan dinamis, menekankan pada pencapaian **tujuan bersama** melalui negosiasi, interaksi, dan berbagi makna di antara anggota kelompok. Strukturnya berpusat pada komunitas praktik (*Communities of Practice*) yang dinamis (Fernandez-Nieto et al., 2024). |
| **Teknologi (Technology)** | Sering kali menggunakan alat-alat standar yang tersedia di *Learning Management Systems* (LMS), seperti forum diskusi, wiki sederhana, atau fitur kuis kelompok (Sidi et al., 2022; Gan et al., 2015). | Mengintegrasikan teknologi canggih seperti **AI dan *machine learning*** untuk personalisasi, analitik pembelajaran (*learning analytics*), dan rekomendasi konten yang adaptif (Sajja et al., 2025; Wu et al., 2024; Ouyang & Wu, 2023). |
| **Contoh Aplikasi** | Tugas terstruktur seperti *Think-Pair-Share* (Samaila et al., 2025), kuis kelompok, atau proyek di mana setiap anggota bertanggung jawab atas bagian tertentu dan dinilai secara individual. | Mahasiswa bersama-sama membuat **artefak pengetahuan** (misalnya, peta konsep, video, esai kolaboratif) yang kemudian dinilai oleh rekan sejawat (*peer-review*) menggunakan rubrik berbasis VCC. Mahasiswa berperan sebagai **produsen konten** (Moro et al., 2024). |
| **Peran Dosen** | Bertindak sebagai **perancang dan manajer pembelajaran**; merancang tugas, membentuk kelompok, dan memantau kemajuan individu secara terstruktur. | Berperan sebagai **fasilitator, mentor, atau *co-inquirer*** (rekan peneliti) yang memandu proses konstruksi pengetahuan, bukan sebagai satu-satunya sumber informasi (Adekola et al., 2017; Robayo-Pinzon et al., 2024). |
