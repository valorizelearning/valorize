# Glosarium

Memahami Bahasa Inovasi * CKM-SE (Collaborative Knowledge Management System-Smart Engineering): Cetak biru operasional dan artefak cerdas yang mengimplementasikan filosofi TISE-VALORIZE.

* Homocordium: Pilar TISE yang merepresentasikan kecerdasan manusia: etika, kreativitas, nilai, dan kebijaksanaan. Kompas moral sistem.
* PUDAL Engine: Mesin kognitif berbasis AI yang mempersonalisasi jalur belajar secara dinamis (Perceive, Understand, Decision-Making, Act-Response, Learning-Evaluating).
* PSKVE Engine: Mesin manajemen nilai yang mengukur kontribusi holistik dalam 5 dimensi (Product, Service, Knowledge, Value, Environmental).
* TISE (Triune-Intelligence Smart Engineering): Paradigma rekayasa yang menyeimbangkan Kecerdasan Manusia (Homocordium), Buatan (Homologos), dan Alamiah.
* VALORIZE: Kerangka pedagogis inti yang berfokus pada pembentukan identitas profesional (becoming) melalui simulasi, Peta Pengetahuan, dan Knowledge Marketplace.
* VCC (Value Co-Creation): Proses di mana nilai diciptakan secara kolaboratif oleh semua pemangku kepentingan (mahasiswa, dosen, komunitas), bukan hanya ditransfer dari dosen ke mahasiswa.
* VOE (Value-Oriented Education): Pendekatan pendidikan yang menempatkan penciptaan nilai sebagai tujuan utamanya, melampaui pencapaian akademik semata.

### Glosarium Proyek: Istilah Kunci dan Definisi

Tabel ini menyediakan daftar lengkap istilah dan konsep fundamental yang digunakan dalam pengembangan dan validasi kerangka kerja VALORIZE dan sistem CKM-SE.

| Istilah | Definisi Lengkap | Konteks & Signifikansi |
| :--- | :--- | :--- |
| **Kerangka Kerja Utama** |
| **CKM-SE** (Collaborative Knowledge Management-Smart Engineering) | Platform pembelajaran digital yang mengintegrasikan AI, *collaborative tools*, dan kerangka kerja berorientasi nilai untuk memfasilitasi pembelajaran personal dan kolaboratif dalam pendidikan tinggi [ ]. Ini adalah *Smart Artifact* utama yang dikembangkan dalam penelitian ini [ ]. | CKM-SE adalah implementasi konkret dan operasional dari filosofi TISE dan VALORIZE. Ini bukan sekadar LMS, melainkan ekosistem rekayasa untuk membentuk identitas profesional [ ]. |
| **VALORIZE** | Paradigma pembelajaran transformatif yang merupakan akronim dari **VAL**ue (Nilai), **CO**llabo**RA**tion (Kolaborasi), **I**ntelligence (Kecerdasan Buatan), dan Personali**ZE**d (Personalisasi) [ ] [ ]. | VALORIZE adalah pendekatan pedagogis yang mendasari CKM-SE. Tujuannya adalah menggeser fokus pendidikan dari "mengetahui" (*knowing*) atau penguasaan konten, menjadi "menjadi" (*becoming*) atau pembentukan sosok, karakter, dan pola pikir rekayasawan profesional [ ] [ ]. |
| **TISE** (Triune-Intelligence Smart Engineering) | Sebuah paradigma arsitektural dan filosofi rekayasa yang bertujuan menyeimbangkan dan menyinergikan tiga jenis kecerdasan: Manusia (*Homocordium*), Buatan (*Homologos*), dan Alamiah (*Natural Intelligence*) [ ] [ ]. | TISE adalah landasan filosofis dan arsitektural untuk CKM-SE [ ]. Tujuannya adalah merancang sistem cerdas yang tidak hanya efisien secara teknis, tetapi juga humanis, selaras dengan nilai, dan bertanggung jawab secara holistik [ ] [ ]. |
| **Komponen Arsitektur & Mesin** |
| **PUDAL Engine** | Mesin kognitif berbasis AI yang menjalankan siklus operasional: **P**erceive (Merasakan), **U**nderstand (Memahami), **D**ecision-Making (Membuat Keputusan), **A**ct-Response (Bertindak-Merespons), dan **L**earning-Evaluating (Belajar-Mengevaluasi) [ ] [ ]. | Berfungsi sebagai "otak" kognitif dari CKM-SE yang memungkinkan personalisasi dan adaptasi dinamis [ ]. Siklus ini memastikan sistem terus belajar dan berevolusi untuk memenuhi kebutuhan pembelajar secara optimal [ ]. |
| **PSKVE Engine** | Mesin manajemen nilai yang mengukur dan mengoptimalkan penciptaan nilai holistik melalui lima dimensi: **P**roduct (Produk), **S**ervice (Layanan), **K**nowledge (Pengetahuan), **V**alue (Nilai), dan **E**nvironmental (Lingkungan) [ ] [ ]. | Berfungsi sebagai "jantung" dari CKM-SE yang mengubah konsep VCC dari abstrak menjadi proses rekayasa yang terukur [ ]. Ini memastikan bahwa nilai yang diciptakan bersifat multi-dimensi, tidak hanya nilai akademik [ ]. |
| **Core Engine** (Mesin Inti) | Fondasi dari setiap kemampuan artefak cerdas. Ini adalah komponen yang bertanggung jawab atas fungsi dasar dan konversi energi menjadi kerja (misalnya, komputasi atau gerak) [ ] [ ]. | Dalam CKM-SE, ini adalah platform KMS fundamental yang menjadi dasar bagi PUDAL dan PSKVE Engine. Tanpa *Core Engine* yang kuat, fungsi cerdas lainnya tidak dapat beroperasi [ ] [ ]. |
| **Knowledge Marketplace** (Pasar Pengetahuan) | Sistem asesmen dinamis yang menyerupai ekonomi mikro profesional. Dosen bertindak sebagai "Pencipta Permintaan" dan mahasiswa sebagai "Pencipta Nilai" yang memproduksi artefak pengetahuan untuk "dibeli" menggunakan mata uang digital bertingkat [ ] [ ]. | Ini adalah mekanisme utama untuk mendorong motivasi intrinsik. Mengubah dinamika penilaian dari kewajiban menjadi kontribusi, sejalan dengan Teori Penentuan Diri (SDT) [ ] [ ]. |
| **Konsep Pedagogis & Pembelajaran** |
| **VCC** (Value Co-Creation) | Proses kolaboratif antara mahasiswa, dosen, dan pemangku kepentingan lainnya untuk menciptakan nilai pembelajaran yang saling menguntungkan dan terukur, di mana hasilnya lebih besar dari jumlah kontribusi individu [ ] [ ]. | VCC adalah tujuan operasional utama dari CKM-SE. Keberhasilannya diukur secara kuantitatif melalui PSKVE Engine dan VCC Index [ ]. |
| **VOE** (Value-Oriented Education) | Pendekatan pedagogis yang menempatkan penciptaan nilai (akademik, sosial, ekonomi, dll.) sebagai tujuan inti dalam desain dan implementasi pembelajaran, melampaui sekadar transfer pengetahuan [ ] [ ]. | VOE adalah filosofi pendidikan yang lebih luas yang ingin diwujudkan oleh CKM-SE. VCC adalah cara untuk mengoperasionalkan VOE di dalam kelas [ ]. |
| **Knowledge Maps** (Peta Pengetahuan) | Instrumen pedagogis visual yang dirancang untuk membuat proses berpikir ahli menjadi eksplisit. Mahasiswa menggunakannya untuk menstrukturkan pengetahuan dan memetakan rute pemecahan masalah [ ] [ ]. | Alat kognitif utama dalam VALORIZE. Sesuai dengan Teori Beban Kognitif (CLT), alat ini mengurangi beban kognitif asing (*extraneous load*) dan membantu mahasiswa membangun skema mental yang kuat [ ] [ ]. |
| **PL** (Personalized Learning) | Pendekatan pendidikan yang menyesuaikan pengalaman belajar dengan kebutuhan, minat, kecepatan, dan gaya belajar masing-masing mahasiswa [ ]. | Di CKM-SE, PL diaktifkan oleh PUDAL Engine, yang memprediksi dan merekomendasikan jalur pembelajaran yang optimal untuk setiap individu [ ]. |
| **CL** (Collaborative Learning) | Pendekatan pedagogis di mana mahasiswa bekerja bersama dalam kelompok untuk mencapai tujuan pembelajaran bersama, saling mendukung, dan membangun pemahaman kolektif [ ]. | CKM-SE memfasilitasi CL melalui fitur seperti forum, tinjauan sejawat (*peer review*), dan penggunaan kembali artefak oleh komunitas [ ]. |
| **Prinsip Filosofis & Desain** |
| **Triune Intelligence** (Kecerdasan Tripartit) | Konsep puncak dalam TISE yang menyatakan bahwa sistem cerdas sejati harus menyinergikan tiga bentuk kecerdasan: Manusia, Buatan, dan Alamiah [ ] [ ]. | Ini adalah fondasi yang memastikan solusi rekayasa tidak hanya kuat secara komputasi, tetapi juga bijaksana, etis, dan membumi [ ]. |
| ***Homocordium*** | Representasi dari Kecerdasan Manusia, yang mencakup nilai-nilai, etika, empati, kreativitas, dan tujuan spiritual. Berasal dari *Homo* (manusia) dan *Cor* (hati) [ ] [ ]. | Berfungsi sebagai "hati" dan kompas moral dari sistem TISE. Ini adalah sumber arahan etis yang memandu kekuatan Kecerdasan Buatan [ ] [ ]. |
| ***Homologos*** | Representasi dari Kecerdasan Buatan (AI), yang mencakup kekuatan komputasi, penalaran logis, dan kemampuan belajar mesin [ ] [ ]. | Berfungsi sebagai "otak logis" dari sistem TISE. Dalam kerangka ini, AI adalah mitra yang kuat, bukan tuan [ ]. |
| **Natural Intelligence** (Kecerdasan Alamiah) | Pengakuan bahwa alam memiliki "kecerdasan" inherenâ€”prinsip, hukum fisika, dinamika ekosistem, dan batasan sumber daya yang harus dihormati [ ] [ ]. | Memberikan batasan realistis dan tujuan keberlanjutan. Memastikan rekayasa yang bijaksana bekerja selaras dengan alam, bukan melawannya [ ]. |
| **Smart Artifact** | Produk atau sistem yang dirancang menggunakan *Design Science Research* yang memiliki karakteristik spesifik seperti Cerdas, Kuat, dan Metodis, untuk memecahkan masalah praktis [ ] [ ]. | CKM-SE secara konseptual didefinisikan sebagai *Smart Artifact*, yaitu artefak utama yang dihasilkan dan divalidasi oleh penelitian ini [ ]. |
| **Metodologi & Evaluasi** |
| **ASTF** | Kerangka kerja hierarkis untuk mengelola kompleksitas rekayasa dan evaluasi, terdiri dari empat lapisan: **A**pplication, **S**ystem, **T**echnology, dan **F**undamental Research [ ] [ ]. | Digunakan untuk menstrukturkan taksonomi ilmu rekayasa baru dan sebagai dasar kerangka evaluasi berlapis untuk CKM-SE, memastikan validasi di setiap tingkatan [ ] [ ]. |
| **PICOC** | Kerangka kerja untuk merancang studi empiris berbasis bukti, terdiri dari: **P**opulation, **I**ntervention, **C**omparison, **O**utcome, dan **C**ontext [ ] [ ]. | Digunakan bersama ASTF untuk merancang protokol validasi yang ketat bagi CKM-SE, memastikan setiap klaim dampak dapat dipertanggungjawabkan secara ilmiah [ ]. |
| **DSRM** (Design Science Research Methodology) | Filosofi dan metodologi penelitian yang berfokus pada penciptaan dan evaluasi artefak untuk memecahkan masalah praktis dan menghasilkan kontribusi teoretis [ ] [ ]. | DSRM adalah metodologi inti yang memandu seluruh siklus penelitian, mulai dari identifikasi masalah hingga desain, pengembangan, dan evaluasi CKM-SE [ ]. |
