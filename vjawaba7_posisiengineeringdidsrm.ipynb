# Posisi Engineering di DSRM
## DSRM
*Design Science Research Methodology* (DSRM) (Peffers et al., 2007, p. 3) adalah metodologi yang mapan untuk penelitian yang bertujuan untuk merancang, memvalidasi dan mengevaluasi artefak inovatif secara sistematis (Proses bagaimana kita meneliti). DSRM menjawab pertanyaan "bagaimana" sebuah artefak inovatif (seperti CKMS-SE) dikembangkan dan divalidasi secara ilmiah.

### Adaptasi DSRM
Kebaruan metodologis utama dari penelitian ini adalah adaptasi DSRM dengan mengintegrasikannya dengan empat level desain instruksional dari Romiszowski. Integrasi ini memastikan bahwa setiap fase dalam siklus DSRM (mulai dari identifikasi masalah hingga evaluasi) selaras dengan prinsip-prinsip desain pembelajaran yang sistematis

DSRM menyediakan proses yang terdiri dari enam fase/langkah: 

1. Identifikasi Masalah
2. Definisi Tujuan
3. Desain dan Pengembangan
4. Demonstrasi
5. Evaluasi 
6. Komunikasi. 

* Pada tahap Definisi Tujuan (DSRM), level 1 Romiszowski (Perencanaan Kursus) digunakan untuk menetapkan tujuan VOE dan VCC secara keseluruhan.
* Pada tahap Desain dan Pengembangan Artefak (DSRM), level 1-4 Romiszowski (dari Perencanaan Kurikulum hingga Desain Latihan) digunakan untuk merancang arsitektur sistem dan pengalaman belajar

Selain itu, fase awal DSRM (Identifikasi Masalah) diperkuat dengan Systematic Literature Review (SLR) mengikuti pedoman dari Kitchenham & Charters (2007) untuk mengidentifikasi celah penelitian secara sistematis.

## Alternatif Metode
Meskipun DSRM adalah metode utama, beberapa metode penelitian lain dapat dianggap sebagai alternatif atau komplementer, meskipun kurang sesuai untuk tujuan penelitian ini:

* Action Research: Metode ini berfokus pada pemecahan masalah praktis dalam konteks organisasi tertentu melalui intervensi partisipatif. Namun, tujuannya kurang berfokus pada penciptaan artefak yang dapat digeneralisasi seperti DSRM  
* Studi Kasus: Dapat digunakan untuk evaluasi mendalam, tetapi kurang menyediakan kerangka kerja untuk proses desain dan pengembangan artefak itu sendiri  
* Eksperimen Murni: Sulit dilakukan dalam konteks pendidikan nyata karena kendala etis dan praktis dalam melakukan randomisasi penuh. Oleh karena itu, penelitian ini memilih studi kuasi-eksperimental sebagai metode evaluasi pada fase akhir  (Johan, n.d., p. 4)

Action Research dipertimbangkan tetapi ditolak karena kurang fokus pada pembangunan artefak teknologi yang reusable. Pure Experimental ditolak karena terlalu kaku untuk pengembangan sistem yang iteratif. DSRM adalah pilihan terbaik untuk riset yang menghasilkan produk (artefak).


## SE
Rekayasa Cerdas (*Smart Engineering*), khususnya paradigma TISE adalah Paradigma Desain-nya (Filosofi apa yang kita bangun). Jadi, TiSE adalah fondasi filosofis dan teknis yang menjadi panduan dalam merancang artefak tersebut. TISE memberikan kerangka kerja tentang "apa" yang harus diwujudkan dalam artefak, yaitu sistem yang mampu menyinergikan kecerdasan manusia, buatan, dan alamiah untuk menciptakan nilai holistik.

## Posisi
Posisi: Kita menggunakan metodologi DSRM untuk membangun artefak yang berlandaskan prinsip Smart Engineering. Engineering terjadi pada tahap ke-3 DSRM (Desain & Pengembangan) di mana kita merekayasa PUDAL dan PSKVE Engine.

Secara sederhana, DSRM adalah peta jalan proses penelitiannya, sedangkan Rekayasa Cerdas (TISE) adalah kompas filosofis yang mengarahkan isi dan tujuan dari setiap langkah di peta jalan tersebut. Misalnya, pada fase "Identifikasi Masalah" DSRM, lensa Rekayasa Cerdas memastikan masalah didefinisikan bukan hanya sebagai masalah teknis, tetapi sebagai kesenjangan nilai (*value gaps*)
