# Metodologi Penelitian

## Desain Studi dan Metodologi

Pengembangan dan evaluasi CKMS-SE mengikuti Design Science Research Methodology (DSRM) yang diartikulasikan oleh Peffers et al. DSRM dipilih karena hasil riset adalah artefak berbasis solusi—sistem CKMS-SE. Studi empiris mengadopsi desain kuasi-eksperimental dengan kelompok kontrol (n=35) dan (n=52) dan kelompok intervensi (n=56), dilakukan selama semester 8 minggu. Populasi terdiri dari 143 mahasiswa sarjana rekayasa yang terdaftar di mata kuliah Probabilitas dan Statistik.

## Komponen Intervensi dan Pemetaan ASTF

Kelompok intervensi menggunakan metode belajar kolaboratif yang distrukturkan sebagai Simulasi Profesi berbasis prinsip TISE-VALORIZE 2.0. Semua artefak utama dikelola menggunakan Git/GitHub. Intervensi terdiri dari tiga sistem terintegrasi:
1. Knowledge Marketplace: Dosen memposting “kebutuhan belajar” mingguan (Sprint Goals).Mahasiswa secara sukarela menciptakan artefak sebagai respons, dihargai dengan Portofolio Nilai PSKVE multi-aset.
2. Knowledge Maps: Alat visualisasi berbantuan AI memandu mahasiswa dalam membuat artefak sesuai Taksonomi Multi-Graf formal (misalnya, DAGs, Weighted Graphs).
3. Socratic AI Coaching: Integrasi API memberikan scaffolding real-time dengan mengajukan pertanyaan klarifikasi tanpa memberikan jawaban langsung, mendukung otonomi mahasiswa (Homodeus melayani Homocordium).

## Instrumen Pengukuran

Metrik dikalibrasi untuk secara eksplisit mengukur keberhasilan pergeseran filosofis dari mengetahui ke menjadi sosok: 
1. Homocordium (Afektif/Identitas): Diukur menggunakan Intrinsic Motivation Inventory dan Learning and Professional Identity in Practice Scale (LPIPS) 
2. Homodeus (Kognitif/Transfer): Diukur melalui penilaian Transfer Learning dengan kesepakatan penilai independen yang kuat.
3. Penciptaan Nilai Holistik (PSKVE): Dinilai melalui metrik platform, termasuk tingkat kreasi artefak, frekuensi peer review, dan, yang terpenting, penggunaan kembali artefak oleh cohort berikutnya.



# Research Methodology

## Study Design and Methodology

The development and evaluation of CKMS-SE followed the Design Science Research Methodology (DSRM) articulated by Peffers et al. DSRM was chosen because the research outcome is a solution-based artifact—the CKMS-SE system. The empirical study adopted a quasi-experimental design with a control group (n=35 and n=52) and an intervention group (n=56), conducted over an 8-week semester. The population consists of 143 undergraduate engineering students enrolled in the Probability and Statistics course.

## Intervention Components and Mapping of ASTF

The intervention group used a collaborative learning method structured as a Profession Simulation based on the TISE-VALORIZE 2.0 principles. All major artefacts were managed using Git/GitHub. The intervention consisted of three integrated systems: 1. Knowledge Marketplace: Lecturers posted weekly "learning needs" (Sprint Goals). Students voluntarily create artefacts as a response, which are rewarded with a multi-asset PSKVE Value Portfolio.
2. Knowledge Maps: AI-assisted visualisation tools guide students in creating artefacts according to formal Multi-Graph Taxonomies (e.g., DAGs, Weighted Graphs). 3. Socratic AI Coaching: API integration provides real-time scaffolding by asking clarifying questions without giving direct answers, supporting student autonomy (Homodeus serves Homocordium).

## Measurement Instruments

Metrics were calibrated to explicitly measure the success of the philosophical shift from knowing to being: 1. Homocordium (Affective/Identity): Measured using the Intrinsic Motivation Inventory and the Learning and Professional Identity in Practice Scale (LPIPS) 2. Homodeus (Cognitive/Transfer): Measured through Transfer Learning assessments with strong independent rater agreement.
3. Holistic Value Creation (PSKVE): Assessed through platform metrics, including the rate of artefact creation, frequency of peer review, and, most importantly, the reuse of artefacts by subsequent cohorts.
