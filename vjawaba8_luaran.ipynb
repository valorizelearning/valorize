# Pedoman (Luaran Praktis)

Penelitian ini tidak hanya menghasilkan temuan teoretis, tetapi juga luaran praktis berupa pedoman yang dapat diterapkan oleh institusi pendidikan tinggi lainnya. Pedoman yang diharapkan meliputi:

* Buku Panduan Implementasi (*Implementation Handbook*): Sebuah panduan praktis yang merinci langkah-langkah untuk mengadopsi dan mengimplementasikan kerangka kerja CKMS-SE di lingkungan perguruan tinggi.
* Materi Pelatihan untuk Dosen: Modul dan materi pelatihan yang dirancang untuk membantu dosen memahami filosofi VALORIZE, protokol standar tentang cara mengadopsi platform CKMS-SE secara efektif dalam pengajaran mereka.
* Rubrik VCC Index: Rubrik penilaian lima dimensi (PSKVE) yang terstruktur dan tervalidasi, yang dapat digunakan oleh institusi lain untuk mengukur penciptaan nilai bersama (VCC) dari artefak pembelajaran mahasiswa secara kuantitatif.
