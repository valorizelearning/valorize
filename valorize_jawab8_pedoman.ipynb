### Pedoman Berlapis dalam Penelitian dan Implementasi CKM-SE

Kerangka kerja CKM-SE tidak dikembangkan secara acak, melainkan dipandu oleh tiga tingkatan pedoman yang memastikan prosesnya rigor, hasilnya koheren secara filosofis, dan implementasinya efektif.

---

#### 1. Pedoman Metodologis: *Design Science Research* (DSRM)

Ini adalah pedoman tingkat tertinggi yang mengatur **bagaimana penelitian ini harus dilakukan secara ilmiah**. Penelitian ini secara eksplisit mengadopsi tujuh pedoman kritis untuk Riset Sains Desain yang dirumuskan oleh Hevner et al. (2004), yang memastikan keketatan dan relevansi penelitian [ ] [ ].

| No. | Pedoman DSRM | Deskripsi dalam Konteks Penelitian CKM-SE |
| :-- | :--- | :--- |
| **1.** | **Desain sebagai Artefak** | Menghasilkan artefak yang dapat diterapkan, yaitu sistem **CKM-SE** sebagai sebuah *Smart Artefact* yang fungsional [ ] [ ]. |
| **2.** | **Relevansi Masalah** | Mengatasi masalah nyata dan penting di pendidikan tinggi, yaitu kesenjangan antara tuntutan era AI dan model pembelajaran konvensional [ ] [ ]. |
| **3.** | **Evaluasi Desain** | Melakukan evaluasi rigor terhadap efektivitas artefak melalui metode kuantitatif (studi quasi-eksperimental) dan kualitatif [ ] [ ]. |
| **4.** | **Kontribusi Riset** | Menghasilkan kontribusi baru berupa kerangka kerja TISE-VALORIZE, artefak CKM-SE, dan bukti empiris efektivitasnya [ ] [ ]. |
| **5.** | **Keketatan Riset** | Menggunakan metodologi yang ketat dalam pengembangan (DSRM) dan evaluasi (PICOC-ASTF) [ ] [ ]. |
| **6.** | **Desain sebagai Proses Pencarian** | Mengakui bahwa desain adalah proses iteratif. Kerangka kerja ini terus disempurnakan berdasarkan temuan dari setiap siklus DSRM [ ]. |
| **7.** | **Komunikasi Riset** | Mengkomunikasikan hasil secara efektif kepada audiens akademis (melalui publikasi jurnal) dan praktisi (melalui panduan implementasi) [ ] [ ]. |

#### 2. Pedoman Filosofis & Arsitektural: Prinsip Inti TISE & VALORIZE

Ini adalah pedoman yang menjawab pertanyaan **"Sistem seperti APA yang harus dibangun?"**. Pedoman ini membentuk jiwa dan arsitektur dari CKM-SE.

*   **Pergeseran dari "Mengetahui" ke "Menjadi" (*From Knowing to Becoming*)**: Pedoman filosofis utama adalah mengubah tujuan pendidikan dari sekadar penguasaan konten teknis (*knowing*) menjadi pembentukan karakter, pola pikir, dan identitas profesional (*becoming*) yang tahan terhadap otomasi AI [ ] [ ] [ ].
*   **Sinergi Tiga Kecerdasan (Prinsip TISE)**: Sistem harus dirancang untuk menyeimbangkan secara harmonis antara **Kecerdasan Manusia** (*Homocordium* sebagai kompas moral), **Kecerdasan Buatan** (*Homologos* sebagai mesin optimasi), dan **Kecerdasan Alam** (sebagai batasan realitas) [ ] [ ] [ ].
*   **Penciptaan Nilai Holistik (Prinsip PSKVE)**: Pedoman desain arsitektural mengharuskan sistem untuk dapat mengukur dan mengoptimalkan nilai secara multi-dimensi—meliputi Produk, Layanan, Pengetahuan, Persepsi Nilai, dan Lingkungan (*Product, Service, Knowledge, Value, Environmental*)—bukan hanya nilai akademis [ ] [ ].
*   **Siklus Kognitif Adaptif (Prinsip PUDAL)**: Sistem harus memiliki "mesin kognitif" internal yang memungkinkannya belajar dan beradaptasi secara terus-menerus melalui siklus *Perceive, Understand, Decision-Making, Act-Response, Learning-Evaluating* [ ] [ ].

---

#### 3. Pedoman Implementasi Praktis: Aturan Main di dalam Kelas

Ini adalah pedoman operasional yang menjawab pertanyaan **"BAGAIMANA mahasiswa dan dosen berinteraksi di dalam ekosistem CKM-SE?"**.

*   **Pedagogi Inti adalah "Simulasi Profesi"**: Seluruh aktivitas di kelas dibingkai sebagai sebuah simulasi lingkungan profesional. Dosen berperan sebagai "Arsitek Sistem" atau klien, sementara mahasiswa berperan sebagai "Insinyur Junior" [ ] [ ].
*   **Ekonomi Pengetahuan melalui *Knowledge Marketplace***:
    1.  **Dosen Mengiklankan Kebutuhan**: Dosen mem-posting "iklan" berisi kebutuhan akan artefak pengetahuan (misalnya, Peta Pemecahan Masalah untuk topik sulit) [ ].
    2.  **Mahasiswa Menciptakan Nilai**: Mahasiswa secara sukarela merespons dengan menciptakan artefak tersebut, yang mendorong otonomi mereka [ ].
    3.  **Transaksi Berbasis Kualitas**: Artefak berkualitas akan "dibeli" oleh dosen menggunakan mata uang digital berjenjang (Uang, Emas, Platinum, Berlian) yang dipetakan ke Taksonomi Bloom [ ] [ ].
    4.  **Publikasi sebagai Aset Komunitas**: Artefak yang "terbeli" dipublikasikan dan menjadi aset pembelajaran bagi angkatan selanjutnya, menciptakan legasi dan rasa kepemilikan [ ].
*   **Validasi Berbasis Bukti di Setiap Lapisan (PICOC-ASTF)**: Setiap kontribusi dan klaim, baik oleh mahasiswa dalam tugasnya maupun oleh peneliti dalam risetnya, harus divalidasi secara sistematis menggunakan kerangka kerja PICOC (Population, Intervention, Comparison, Outcome, Context) yang diterapkan pada setiap lapisan arsitektur ASTF (Application, System, Technology, Fundamental) [ ] [ ].

Secara ringkas, pedoman-pedoman ini memastikan bahwa proyek CKM-SE tidak hanya canggih secara teknis, tetapi juga kokoh secara metodologis, mendalam secara filosofis, dan efektif secara pedagogis.
