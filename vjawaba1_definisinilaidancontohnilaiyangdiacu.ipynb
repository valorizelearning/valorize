# Definisi Nilai dan Contoh Nilai yang Diacu

## Definisi Nilai
Dalam konteks penelitian ini, **"Nilai" (*Value*)** tidak didefinisikan secara umum/sempit sebagai nilai ekonomi (uang) atau nilai akademik (IPK), melainkan didefinisikan melalui dua konsep utama: Pendidikan Berorientasi Nilai (*Value-Oriented Education*/VOE) dan Penciptaan Nilai Bersama (*Value Co-Creation*/VCC), yaitu manfaat nyata yang diciptakan bersama oleh mahasiswa untuk komunitasnya.

* Pendidikan Berorientasi Nilai (***Value-Oriented Education***/VOE) adalah pendekatan pedagogis yang menempatkan penciptaan nilai mencakup aspek akademik, sosial, budaya, ekonomi, dan keberlanjutan sebagai tujuan inti dalam desain, implementasi, dan evaluasi pembelajaran. VOE menggeser fokus dari sekadar transfer pengetahuan menjadi pembentukan karakter dan pola pikir profesional.
* Penciptaan Nilai Bersama (***Value Co-Creation***/VCC) adalah proses kolaboratif antara mahasiswa, dosen, dan pemangku kepentingan lainnya (seperti industri) untuk menciptakan nilai pembelajaran yang saling menguntungkan dan terukur. Ini adalah mekanisme operasional untuk merealisasikan VOE.

Untuk mengoperasionalkan dan mengukur VCC, penelitian ini menggunakan **PSKVE *Engine***, yang mengukur nilai dari artefak pembelajaran (misalnya, laporan, video, simulasi) dalam lima dimensi energi.

## Contoh Nilai yang Diacu
Contoh Nilai (Menggunakan Kerangka PSKVE). Nilai diukur melalui 5 dimensi energi :   

* P - *Product Quality* - Kualitas Produk: Mengukur keunggulan teknis dan profesionalisme artefak yang dihasilkan. **Contoh**: Mahasiswa membuat perhitungan statistik dalam sebuah laporan proyek.
* S - *Service Utility* - Kegunaan Layanan: Mengukur penerapan praktis dan kegunaan langsung dari artefak bagi pengguna akhir (misalnya, komunitas atau angkatan berikutnya). **Contoh**: solusi yang ditawarkan dalam proyek benar-benar memecahkan masalah di dunia nyata misalnya digunakan/diterapkan oleh UMKM Lokal.
* K - *Knowledge Transfer* - Transfer Pengetahuan/Reusabilitas: Mengukur kualitas penyebaran pengetahuan serta potensi artefak untuk digunakan kembali dan diadaptasi oleh orang lain. **Contoh**: Dokumentasi yang sangat jelas sehingga mahasiswa angkatan berikutnya dapat memodifikasinya tanpa memulai dari nol, dapat diadaptasi untuk masalah lain dan seberapa sering artefak tersebut diunduh atau dikutip oleh rekan-rekannya.
* V - *Value-in-Use* - Nilai Guna: Mengukur nilai jangka panjang yang dirasakan oleh mahasiswa (produsen) melalui proses VCC, serta nilai yang dirasakan oleh komunitas. **Contoh**: Mahasiswa melaporkan penguasaan keterampilan baru dan artefak tersebut mengarah pada proses baru di komunitas.
* E - *Environmental/Ethical Impact* - Dampak Etis dan Sosial: Mengukur kepatuhan terhadap pedoman etika, pertimbangan isu sosial/lingkungan, dan keadilan dalam penanganan data. **Contoh**: Analisis data menjamin privasi pelanggan, tidak bias, dan mempertimbangkan prinsip keberlanjutan.


## Kumpulan Data Penelitian dan Referensi Mengenai Nilai (*Value*) dalam Pendidikan Tinggi

Berikut ini merupakan rangkuman komprehensif dari berbagai penelitian yang membahas konsep "nilai" (*value*) dalam konteks Pendidikan Tinggi. Tabel ini mengintegrasikan berbagai perspektif tentang nilai, mulai dari persepsi mahasiswa, desain sistem berbasis nilai, hingga nilai strategis dan ekonomi, yang diidentifikasi dari berbagai sumber penelitian.

| No. | Peneliti (Tahun) | Jenis Nilai yang Dibahas | Deskripsi/Konsep Kunci | Indikator atau Dimensi |
| :-- | :--- | :--- | :--- | :--- |
| 1. | Adikari et al. (2021) | **Nilai Strategis dari Data** (*Strategic Value from Data*) | Nilai diidentifikasi dengan menganalisis opini dan umpan balik mahasiswa dari data (misalnya media sosial) untuk diubah menjadi wawasan yang dapat ditindaklanjuti untuk strategi institusional. | Pengalaman mahasiswa, Kekhawatiran tentang daya saing lulusan (*employability*), Pertumbuhan jumlah mahasiswa, Keunggulan riset, Kemitraan industri. |
| 2. | Dollinger, Lodge, & Coates (2018) | **Nilai Dua Dimensi** (*Co-production & Value-in-Use*) | Nilai merupakan proses ganda: **Co-production** (penciptaan bersama "proposisi nilai" seperti kurikulum) dan **Value-in-Use** (bagaimana mahasiswa secara pribadi menggunakan dan merasakan nilai dari proposisi tersebut). | **Indikator Co-production:** Pembagian pengetahuan, Kesetaraan, Interaksi. <br> **Indikator Value-in-Use:** Pengalaman, Personalisasi, Hubungan. |
| 3. | Hafeez et al. (2022) | **Nilai sebagai Kualitas Pendidikan** (*Value as Educational Quality*) | Nilai didefinisikan melalui berbagai klasifikasi kualitas dari perspektif beragam pemangku kepentingan (mahasiswa, pemberi kerja). Kualitas adalah konsep multifaset yang mencakup tujuan, keunggulan, transformasi, dan akuntabilitas. | **Purposeful** (sesuai visi/misi), **Exceptional** (memenuhi standar yang ketat), **Transformative** (menciptakan perubahan positif pada pembelajar), **Accountable** (bertanggung jawab kepada pemangku kepentingan). |
| 4. | Jamil et al. (2025) | **Nilai Relasional dan Dampak Sosial** (*Relational & Social Impact Value*) | Nilai diciptakan melalui pembangunan hubungan positif, komunitas, dan rasa memiliki (*sense of belonging*). Inisiatif pendidikan juga dapat "didorong oleh nilai" (*value-driven*) untuk menyelesaikan masalah sosial di dunia nyata. | Komunitas, Kolaborasi, Kohesi (dari model 3 C), Kesejahteraan (*well-being*), Keberlanjutan (*sustainability*), Keadilan sosial. |
| 5. | Mohammed et al. (2021) | **Nilai Guna yang Dirasakan dari Sistem Manajemen Pengetahuan (KMS)** | Nilai sebuah sistem diukur melalui kegunaan yang dirasakan oleh pengguna, yang dipengaruhi oleh kualitas informasi, kualitas sistem itu sendiri, dan kualitas layanan yang menyertainya. | Kualitas Informasi (IKQ), Kualitas Sistem (SYQ), Kualitas Layanan (SEQ), Penggunaan (KU), Kepuasan Pengguna (US). |
| 6. | Pereira & Baranauskas (2015) | **Nilai yang Berorientasi Budaya dalam Desain Sistem** | Nilai secara eksplisit dipertimbangkan selama proses desain sistem, dengan fokus pada relevansi budaya dan kebutuhan pemangku kepentingan. Nilai bukanlah hasil sampingan, melainkan persyaratan inti. | Identitas, Percakapan, Meta-komunikasi, Keamanan, Otonomi, Berbagi, Estetika, Kepemilikan, Aksesibilitas. |
| 7. | Prahalad & Ramaswamy (2004) | **Nilai sebagai Pengalaman** (*Value as Experience*) | Konsep fundamental yang menyatakan bahwa lokus penciptaan nilai bergeser dari produk/layanan ke **pengalaman unik dan personal** yang diciptakan bersama (*co-created*) melalui interaksi antara perusahaan (universitas) dan konsumen (mahasiswa). | Dialog, Akses, Transparansi, Pemahaman risiko-manfaat. |
| 8. | Robayo-Pinzon et al. (2024) | **Penciptaan & Penghancuran Nilai Bersama** (*Value Co-creation & Co-destruction*) | Nilai dipersepsikan oleh mahasiswa sebagai sesuatu yang diciptakan secara kolaboratif (misalnya, melalui AI yang mendukung) atau dihancurkan (misalnya, melalui AI yang tidak personal seperti *chatbot*). | **Indikator Penciptaan Nilai:** Optimalisasi aktivitas akademik, Peningkatan hubungan antarmanusia, Pembelajaran kolaboratif, Komunikasi langsung. <br> **Indikator Penghancuran Nilai:** Dehumanisasi pendidikan, Risiko privasi, Hilangnya esensi pendidikan. |
| 9. | Santos et al. (2023) | **Nilai Ekonomi** (Daya Saing Lulusan) | Nilai sebuah program pendidikan tinggi diukur dari kemampuannya untuk mempersiapkan lulusan menghadapi pasar kerja. Ini adalah pandangan makro mengenai *employability* dari perspektif institusi. | Tingkat penyerapan kerja alumni, Kesesuaian pekerjaan dengan jurusan, Tren pasar tenaga kerja, Keunggulan institusional. |
