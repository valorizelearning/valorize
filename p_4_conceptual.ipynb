# Kerangka Konseptual dan Teoretis

Desain CKMS-SE mensintesis teori-teori pedagogis yang sudah mapan dengan filosofi TISE-VALORIZE yang komprehensif untuk menciptakan ekosistem yang koheren.

## Inti Filosofis: Sintesis TISE-VALORIZE
Inti filosofis dari kerangka TISE-VALORIZE adalah pergeseran yang disengaja dari mengetahui (penguasaan konten) menjadi menjadi sosok (pembentukan karakter, pola pikir, dan penilaian profesional).

Reorientasi ini secara strategis merespons perkembangan kecerdasan buatan dan menegaskan bahwa nilai seorang insinyur berlabuh pada atribut manusiawi (Homocordium) yang diwujudkan (etika, kreativitas, empati). Lingkungan pembelajaran TISE-VALORIZE dirancang sebagai "teater yang indah" untuk belajar bagaimana menjadi seorang insinyur.

## Konsep Pendidikan Fundamental (VOE, VCC, KM, CL, PL)
Platform CKMS-SE didasarkan pada integrasi strategis beberapa konsep kunci untuk mengatasi tuntutan Era AI:
1. Value-Oriented Education (VOE): VOE menekankan bahwa pendidikan harus membentuk karakter dan nilai yang bermanfaat bagi individu dan masyarakat, menekankan pembelajaran berbasis etika, tanggung jawab sosial, dan kontribusi komunitas.
2. Value Co-Creation (VCC): VCC melibatkan interaksi antara mahasiswa, dosen, dan teknologi untuk menciptakan pengalaman belajar yang bermakna, memposisikan mahasiswa sebagai partisipan aktif dalam penciptaan pengetahuan bersama.
3. Knowledge Management (KM): KM bertujuan untuk mengorganisasi, menyimpan, dan berbagi pengetahuan secara efisien di lingkungan pendidikan. AI menggeser kebutuhan dari replikasi pengetahuan menjadi pengelolaan pengetahuan.
4. Collaborative Learning (CL): CL menekankan pentingnya kerja sama dalam pembelajaran untuk mencapai pemahaman yang lebih dalam dan keterampilan berpikir kritis.
5. Personalised Learning (PL): PL mengadaptasi pembelajaran berdasarkan gaya belajar, tingkat pemahaman, dan preferensi mahasiswa, dengan AI berperan kunci dalam analisis data dan umpan balik real-time.

## Fondasi Psikologis dan Kognitif (SDT, CLT, CoP)
CKMS-SE mengintegrasikan kerangka psikologis dan kognitif tradisional:

1.  Self-Determination Theory (SDT): Motivasi intrinsik muncul dari kepuasan kebutuhan psikologis: otonomi, kompetensi, dan keterkaitan (relatedness).
2.  Cognitive Load Theory (CLT): Pembelajaran dioptimalkan ketika beban kognitif ekstrinsik diminimalkan.
3.  Communities of Practice (CoP): Pengetahuan muncul melalui partisipasi dan kontribusi yang bermakna pada repositori bersama.

## Mesin Kognitif: Taksonomi Multi-Graf dan PUDAL
Untuk memodelkan proses kognitif ahli secara ketat, VALORIZE 2.0 memformalkan Knowledge Maps menggunakan taksonomi Multi-Graf yang berfungsi sebagai “tata bahasa pengetahuan” untuk pemikiran ahli.

Seluruh proses kognitif distrukturkan oleh siklus iteratif PUDAL (*Perceive, Understand, Decision-making and Planning, Act-Response, Learning-evaluating), yang bertindak sebagai “mesin mikro-evolusioner” untuk penyempurnaan berkelanjutan. Siklus ini dipandu oleh kerangka Triune Intelligence (TI) TISE yang terdiri dari (1) kecerdasan manusia (Homocordium) yang berfokus pada dimensi etika, moral, dan sosial, (2) kecerdasan buatan (Homodeus) yang berfokus pada aspek komputasi, data, dan logika, (3) kecerdasan alamiah (Natural Intelligence) yang berfokus pada hukum fisik, kendala sumber daya, dan ekologi.

# Conceptual and Theoretical Framework

The CKMS-SE design synthesises established pedagogical theories with the comprehensive TISE-VALORIZE philosophy to create a coherent ecosystem.

## Philosophical Core: Synthesis of TISE-VALORIZE The philosophical core of the TISE-VALORIZE framework is the deliberate shift from knowing (content mastery) to becoming (character development, mindset, and professional judgement).

This reorientation strategically responds to developments in artificial intelligence and affirms that an engineer's value is anchored in embodied human attributes (Homocordium) (ethics, creativity, empathy). The TISE-VALORIZE learning environment is designed as a "beautiful theatre" for learning how to become an engineer.

## Fundamental Education Concepts (VOE, VCC, KM, CL, PL) 

The CKMS-SE platform is based on the strategic integration of several key concepts to address the demands of the AI Era: 1. Value-Oriented Education (VOE): VOE emphasises that education should shape character and values that are beneficial for individuals and society, focussing on ethical learning, social responsibility, and community contribution. 2. Value Co-Creation (VCC): VCC involves interaction between students, lecturers, and technology to create meaningful learning experiences, positioning students as active participants in the co-creation of knowledge. 3. Knowledge Management (KM): KM aims to efficiently organise, store, and share knowledge within an educational environment. AI shifts the need from knowledge replication to knowledge management. 4. Collaborative Learning (CL): CL emphasises the importance of cooperation in learning to achieve deeper understanding and critical thinking skills. 5. Personalised Learning (PL): PL adapts learning based on students' learning styles, understanding levels, and preferences, with AI playing a key role in data analysis and real-time feedback.

## Psychological and Cognitive Foundations (SDT, CLT, CoP) CKMS-SE integrates traditional psychological and cognitive frameworks:

1. Self-Determination Theory (SDT): Intrinsic motivation arises from the satisfaction of psychological needs: autonomy, competence, and relatedness. 
2. Cognitive Load Theory (CLT): Learning is optimised when extrinsic cognitive load is minimised.
3. Communities of Practice (CoP): Knowledge emerges through meaningful participation and contributions to a shared repository.

## Cognitive Engine: Multi-Graph Taxonomy and PUDAL To strictly model expert cognitive processes, VALORISE 2.0 formalises Knowledge Maps using a Multi-Graph taxonomy that serves as a "grammar of knowledge" for expert thinking.

The entire cognitive process is structured by the iterative PUDAL cycle (*Perceive, Understand, Decision-making and Planning, Act-Response, Learning-evaluating), which acts as a "micro-evolutionary engine" for continuous improvement. This cycle is guided by the Triune Intelligence (TI) TISE framework, which consists of (1) human intelligence (Homocordium) focussing on ethical, moral, and social dimensions, (2) artificial intelligence (Homodeus) focussing on computational, data, and logical aspects, and (3) natural intelligence focussing on physical laws, resource constraints, and ecology.

The Knowledge Marketplace was redesigned into a multi-asset PSKVE (product, system, knowledge, value, environment) value portfolio, which directly simulates the three pillars of modern engineering value: Economic, Protective, and Social.
