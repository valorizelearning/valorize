### Adaptasi Metodologi DSRM dan Analisis Alternatifnya DSRM dipilih sebagai metodologi utama karena merupakan paradigma yang paling sesuai untuk penelitian yang bertujuan **menciptakan dan mengevaluasi artefak inovatif** untuk memecahkan masalah praktis sambil memberikan kontribusi teoretis [ ] [ ]. Namun, untuk memaksimalkan relevansinya, DSRM diadaptasi secara khusus. --- ### 1. Adaptasi Spesifik DSRM dalam Penelitian CKM-SE Adaptasi ini mengubah DSRM dari sekadar proses rekayasa perangkat lunak menjadi kerangka kerja pengembangan sistem sosio-teknis yang komprehensif. #### a. Integrasi dengan Level Desain Instruksional Romiszowski Ini adalah adaptasi paling fundamental. Setiap fase DSRM dipetakan secara eksplisit ke level desain instruksional yang relevan. Ini memastikan bahwa artefak yang dibangun tidak hanya fungsional secara teknis, tetapi juga kokoh secara pedagogis [ ] [ ]. | Tahap DSRM | Aktivitas Utama | Level Desain Instruksional Terintegrasi | Output yang Dihasilkan |
| :--- | :--- | :--- | :--- |
| **1. Identifikasi Masalah** | Analisis kebutuhan & celah kinerja. | **Analisis Kebutuhan Pelatihan** | Pernyataan masalah yang jelas. |
| **2. Definisi Tujuan Solusi** | Mendefinisikan tujuan fungsional CKM-SE. | **Level 1: Perencanaan Kursus** (Menetapkan tujuan VOE & VCC). | Spesifikasi persyaratan tingkat tinggi. |
| **3. Desain & Pengembangan** | Merancang arsitektur teknis & pengalaman belajar. | **Level 1-4: Dari Kurikulum hingga Desain Latihan** | Arsitektur Sistem CKM-SE dan Prototipe Fungsional. |
| **4. Demonstrasi** | Menerapkan prototipe dalam studi kasus. | **Implementasi Terkendali** | Data penggunaan sistem dari studi kasus. |
| **5. Evaluasi** | Mengukur efektivitas dan dampak CKM-SE. | **Evaluasi Formatif dan Sumatif** | Hasil validasi kualitatif dan kuantitatif. |
| **6. Komunikasi** | Menyebarkan temuan penelitian. | **Diseminasi Hasil** | Disertasi, publikasi jurnal, panduan praktis. | #### b. Pembingkaian dalam Paradigma TISE (Triune-Intelligence Smart Engineering) DSRM menyediakan kerangka **"BAGAIMANA"** proses desain dan evaluasi dilakukan, sementara paradigma TISE memberikan kerangka **"APA"** yang harus diwujudkan [ ]. * **DSRM**: Metodologi proses yang sistematis (Identifikasi → Desain → Evaluasi).
* **TISE**: Fondasi filosofis dan teknis yang menuntut agar artefak yang dihasilkan menyeimbangkan kecerdasan manusia (nilai), buatan (optimasi), dan alam (konteks) [ ]. Dengan demikian, setiap tahap DSRM diinformasikan oleh prinsip-prinsip TISE, memastikan artefak yang dihasilkan "Cerdas" secara holistik, bukan hanya secara teknis [ ]. #### c. Peningkatan Fase Evaluasi dengan Kerangka PICOC-ASTF Fase "Evaluasi" standar dalam DSRM diperluas menjadi **validasi berlapis** yang lebih ketat menggunakan kerangka PICOC-ASTF [ ]. Ini mengubah evaluasi dari sekadar daftar metrik menjadi serangkaian eksperimen terstruktur yang memvalidasi klaim di setiap lapisan: * **F**undamental: Validitas teori (dinilai ahli).
* **T**echnology: Kinerja teknis (akurasi AI, *usability*).
* **S**ystem: Kepuasan pengguna, kualitas kolaborasi.
* **A**pplication: Dampak pada hasil belajar (motivasi, transfer) [ ] [ ]. --- ### 2. Analisis Metode Alternatif Meskipun DSRM adalah pilihan yang paling tepat, metode lain bisa dipertimbangkan. Tabel berikut menjelaskan mengapa metode-metode tersebut pada akhirnya kurang sesuai untuk tujuan penelitian ini. | Metodologi Alternatif | Deskripsi Singkat | Kelebihan | Keterbatasan (Mengapa Tidak Dipilih) |
| :--- | :--- | :--- | :--- |
| **Penelitian Tindakan (*Action Research*)** | Siklus berulang (rencana, tindakan, observasi, refleksi) untuk memperbaiki praktik dalam konteks lokal yang spesifik. | Sangat baik untuk perbaikan berkelanjutan di satu kelas atau sekolah. Fokus pada praktisi. | **Kurang fokus pada pembuatan artefak yang general dan dapat direplikasi.** Tujuannya lebih ke perbaikan situasi lokal, bukan menghasilkan teori desain atau *blueprint* sistem yang dapat diimplementasikan di tempat lain. |
| **Studi Kasus (*Case Study*)** | Investigasi mendalam terhadap satu fenomena atau kasus dalam konteks dunia nyata. | Memberikan pemahaman yang kaya dan mendalam tentang "mengapa" dan "bagaimana" suatu intervensi bekerja dalam konteksnya. | **Bersifat deskriptif/analitis, bukan generatif.** Studi kasus lebih cocok untuk menganalisis sistem yang *sudah ada*, bukan untuk memandu proses *penciptaan* dan *pengembangan* artefak baru dari awal. |
| **Eksperimen Murni (*True Experiment*)** | Melibatkan alokasi acak peserta ke dalam kelompok kontrol dan perlakuan untuk menguji hipotesis kausal. | Standar emas untuk membuat klaim sebab-akibat yang kuat. Kontrol tinggi terhadap variabel perancu. | **Tidak praktis dan seringkali tidak etis dalam konteks pendidikan nyata.** Sulit untuk secara acak menugaskan mahasiswa ke metode pengajaran yang berbeda dalam satu mata kuliah. DSRM lebih fleksibel dengan mengakomodasi **studi kuasi-eksperimental** dalam fase evaluasinya [ ]. | ### Kesimpulan: Mengapa DSRM adalah Pilihan yang Tepat DSRM dipilih karena secara unik menjembatani dua dunia: **relevansi praktis** dan **keketatan ilmiah** [ ]. Metode ini adalah satu-satunya yang secara eksplisit berfokus pada: 1. **Penciptaan Artefak**: Menghasilkan solusi nyata (CKM-SE) sebagai inti kontribusi.
2. **Siklus Ganda**: Secara simultan membangun artefak (siklus relevansi) dan menghasilkan pengetahuan desain yang dapat digeneralisasikan (siklus keketatan) [ ].
3. **Fleksibilitas Evaluasi**: Mampu mengintegrasikan berbagai metode evaluasi, termasuk quasi-eksperimen, untuk mengukur dampak artefak di dunia nyata. Dengan adaptasi yang telah dijelaskan, DSRM menjadi metodologi yang tidak hanya kuat, tetapi juga sangat selaras dengan tujuan filosofis dan pedagogis dari penelitian ini.
