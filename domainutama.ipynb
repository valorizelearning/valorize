# *Systematic Literature Review*

## 1. *Review*

### *Title*
Collaborative Knowledge Management Framework Based on Smart Engineering for Value-Oriented Higher Education
### *Description*
Penelitian ini mengusulkan sebuah Kerangka Kerja Manajemen Pengetahuan Kolaboratif (CKM) yang inovatif untuk menggabungkan Rekayasa Cerdas dengan Manajemen Pengetahuan Kolaboratif di perguruan tinggi. Fokus utamanya bukan sekadar membuat sistem kampus menjadi lebih "pintar" secara digital, tetapi menggunakan kecerdasan sistem tersebut untuk mendukung Pendidikan Berorientasi Nilai yang memfasilitasi pertukaran pengetahuan yang efektif antara dosen dan mahasiswa. Berbeda dengan sistem manajemen kampus konvensional yang hanya berfokus pada transfer pengetahuan. Tujuannya adalah memastikan bahwa integrasi teknologi tidak hanya meningkatkan kompetensi teknis, tetapi juga tetap menjunjung tinggi nilai-nilai kemanusiaan dan etika.

## 2. *Planning*

### *Protocol*

#### *Objectives*
*General Objective* (Tujuan Umum)
Merancang dan mengembangkan kerangka kerja Manajemen Pengetahuan Kolaboratif (CKM) berbasis Rekayasa Cerdas (Smart Engineering) yang mampu mengintegrasikan peningkatan kompetensi teknis dengan penanaman nilai-nilai etika dan kemanusiaan dalam lingkungan pendidikan tinggi.

*Specific Objectives* (Tujuan Khusus)
1. Menganalisis Kesenjangan Sistem (*Gap Analysis*): Menganalisis keterbatasan sistem manajemen pengetahuan konvensional di perguruan tinggi saat ini, khususnya dalam kemampuannya mendukung aspek pendidikan karakter dan transfer nilai (value transfer) daripada sekadar transfer data akademik.

2. Merancang Arsitektur CKM Berbasis Rekayasa Cerdas: Membangun arsitektur sistem CKM yang mengadopsi prinsip *Smart Engineering* untuk memfasilitasi pertukaran pengetahuan yang efektif antara dosen dan mahasiswa. Sistem tidak hanya menilai capaian teknis, tetapi juga mampu mendeteksi, mendorong, dan memvalidasi interaksi yang mencerminkan nilai etika, empati, dan kolaborasi positif.

3. Mengevaluasi Efektivitas Kerangka Kerja: Menguji validitas dan efektivitas kerangka kerja yang diusulkan untuk memastikan bahwa penerapannya mampu meningkatkan kompetensi *hard skill* (teknis) mahasiswa sekaligus memperkuat *soft skill* (nilai kemanusiaan) dibandingkan dengan metode konvensional.
