# Lampiran Teknis: Metodologi & Perhitungan Statistik

Dokumen ini merinci asal-usul data dan prosedur perhitungan statistik yang digunakan untuk memvalidasi efektivitas kerangka kerja **VALORIZE (CKM-SE)**. Analisis dilakukan menggunakan dataset nilai Ujian Tengah Semester (UTS) mata kuliah Probabilitas & Statistika ($N=143$).

---

## Statistik Deskriptif (*Descriptive Statistics*)

Angka-angka ini memberikan gambaran umum tentang pusat data (*central tendency*) dan sebarannya (*dispersion*) untuk setiap kelompok kelas.

### Rata-rata Skor (*Mean*)
Dihitung dengan menjumlahkan seluruh nilai mahasiswa dalam satu kelas, dibagi dengan jumlah mahasiswa ($N$).

* **Rumus:** $\bar{x} = \frac{\sum x_i}{N}$
* **Perhitungan:**
    * **K1 (Kontrol 1):** Total Skor $775 \div 35$ Mahasiswa = **22.14**
    * **K2 (Intervensi):** Total Skor $1.396 \div 56$ Mahasiswa = **24.93**
    * **K3 (Kontrol 2):** Total Skor $1.770 \div 52$ Mahasiswa = **34.04**

### Standar Deviasi (*Standard Deviation*)
Mengukur seberapa jauh variasi nilai mahasiswa dari rata-rata kelasnya.

* **K1 ($SD=12.35$):** Variasi nilai tinggi (kemampuan mahasiswa sangat beragam/timpang).
* **K2 ($SD=12.55$):** Variasi nilai tinggi.
* **K3 ($SD=9.92$):** Variasi nilai paling rendah.
    * *Interpretasi:* Nilai mahasiswa di kelas K3 lebih "seragam" dan konsisten di angka tinggi, menunjukkan metode pengajaran berhasil merata ke hampir seluruh mahasiswa.

---

## Uji Signifikansi: One-Way ANOVA

Uji ini menjawab pertanyaan: *"Apakah perbedaan rata-rata antar kelas ini nyata atau hanya kebetulan?"*

* **F-Statistic ($F$): 13.34**
    * Angka ini adalah rasio antara *variasi antar-kelompok* (perbedaan method) dibagi *variasi dalam-kelompok* (perbedaan individu). Semakin tinggi nilai F, semakin besar dampak metode pengajaran dibanding faktor individu.
* **P-Value ($p$): $4.96 \times 10^{-6}$ (< 0.001)**
    * Nilai probabilitas kesalahan hampir nol.
    * *Kesimpulan:* Kita memiliki keyakinan >99.9% bahwa perbedaan hasil belajar ini **signifikan secara statistik** dan bukan disebabkan oleh faktor acak.

---

## Analisis Lanjut: Post-hoc Tukey HSD

Setelah ANOVA menyatakan "ada beda", uji Tukey menentukan "siapa beda dengan siapa" dengan menghitung selisih rata-rata ($\Delta Mean$).

| Perbandingan | Perhitungan ($\Delta Mean$) | Hasil | Interpretasi |
| :--- | :--- | :--- | :--- |
| **K2 vs K1** | $24.93 - 22.14 = \mathbf{+2.79}$ | Tidak Signifikan ($p>0.05$) | Intervensi awal (K2) mulai menunjukkan tren positif dibanding baseline terendah, meski belum drastis. |
| **K3 vs K2** | $34.04 - 24.93 = \mathbf{+9.11}$ | **Signifikan ($p<0.001$)** | Terdapat *gap* performa yang nyata antara Kontrol Unggulan (K3) dengan Intervensi (K2). |

---

## Ukuran Dampak (*Effect Size*): Cohen's d

Metrik ini adalah standar emas dalam riset pendidikan untuk mengukur **seberapa kuat** dampak sebuah intervensi, terlepas dari jumlah sampelnya.

**Rumus:** $d = \frac{\bar{x}_1 - \bar{x}_2}{SD_{pooled}}$

### Dampak Intervensi terhadap Baseline (K2 vs K1)
* Selisih Rata-rata: $2.79$
* Standar Deviasi Gabungan ($SD_{pooled}$): $\approx 12.45$
* **Perhitungan:** $2.79 \div 12.45 = \mathbf{0.22}$
* **Kategori:** ***Small Effect*** (Dampak Kecil).
* *Makna:* Metode kolaboratif di K2 memberikan dampak positif awal, namun masih perlu dioptimalkan.

### Perbandingan dengan Performa Puncak (K3 vs K2)
* Selisih Rata-rata: $9.11$
* Standar Deviasi Gabungan ($SD_{pooled}$): $\approx 11.3$
* **Perhitungan:** $9.11 \div 11.3 = \mathbf{0.80}$
* **Kategori:** ***Large Effect*** (Dampak Besar).
* *Makna:* Perbedaan antara kelas K3 dan K2 sangat tajam. Ini menjadi temuan penting untuk evaluasi selanjutnya: faktor apa di K3 (apakah *input* mahasiswa atau metode instruktur) yang menyebabkan skor sangat tinggi ini?

---

**Ringkasan Akhir:**
Analisis data membuktikan bahwa penerapan awal VALORIZE (K2) berhasil mengangkat performa di atas batas bawah (*baseline*) kelas konvensional (K1). Namun, tantangan berikutnya adalah meningkatkan efektivitasnya agar dapat menyamai atau melampaui performa kelas unggulan (K3).
