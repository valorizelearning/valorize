# Studi Kasus
## **Dari Kebingungan Menjadi Kreator: Perjalanan Seorang Mahasiswa dalam Ekosistem VALORIZE**

Untuk memahami dampak nyata dari CKM-SE, mari kita ikuti perjalanan **Budiarti**, seorang mahasiswa fiktif di tahun ketiga program studi Teknik Elektro yang mengambil mata kuliah wajib, **Probabilitas dan Statistika**.

#### **Tahap 1: Tantangan Awal (Paradigma Lama)**

Di awal semester, perkuliahan berjalan secara konvensional @23. Dosen menjelaskan konsep abstrak seperti "Teorema Bayes" di depan kelas. Budiarti merasa kesulitan; ia memahami rumus dasarnya tetapi gagal menghubungkannya dengan masalah dunia nyata. Ia merasa hanya menjadi penerima pasif informasi, dan motivasinya mulai menurun. Ini adalah cerminan dari kesenjangan pedagogis yang ingin diatasi oleh VALORIZE: fokus pada *mengetahui* (*knowing*) daripada *menjadi* (*becoming*) seorang pemecah masalah @8 @50.

#### **Tahap 2: Intervensi CKM-SE (Titik Balik)**

Minggu berikutnya, dosen—yang kini berperan sebagai **Arsitek Sistem** @10—memperkenalkan ekosistem CKM-SE. Ia memposting sebuah "Sprint Goal" di **Knowledge Marketplace**: "Buatlah artefak pengetahuan yang dapat menjelaskan Teorema Bayes kepada audiens non-teknis dengan menggunakan contoh kasus dari kehidupan sehari-hari." @10 @196

Budiarti, yang sekarang berperan sebagai **"Associate TISE Systems Engineer"** @11, memutuskan untuk menerima tantangan ini.

1.  **Interaksi dengan PUDAL Engine:** Saat Budiarti mulai meneliti, **PUDAL Engine** @135 menganalisis profil belajarnya (yang teridentifikasi sebagai pembelajar visual) dan riwayat interaksinya @116. Sistem kemudian merekomendasikan serangkaian sumber belajar yang dipersonalisasi: bukan hanya teks, tetapi juga video visualisasi, simulasi interaktif, dan studi kasus yang relevan @119.

2.  **Penciptaan Artefak:** Terinspirasi oleh sumber daya tersebut, Budiarti memutuskan untuk tidak hanya menulis rangkuman. Ia merancang sebuah **infografis interaktif** yang menjelaskan Teorema Bayes menggunakan contoh diagnosis medis. Proses ini memaksanya untuk benar-benar memahami konsep secara mendalam agar bisa menyederhanakannya—sebuah pergeseran dari hafalan ke model mental yang kuat @68. Ia menggunakan **Git/GitHub** untuk mengelola versi drafnya, sebuah praktik profesional standar industri @18.

3.  **Penilaian di Knowledge Marketplace:** Budiarti mengunggah infografisnya ke *Knowledge Marketplace*. Karyanya kemudian ditinjau oleh rekan-rekannya dan dievaluasi oleh dosen menggunakan **rubrik PSKVE Engine** @128.

| Dimensi PSKVE | Evaluasi Artefak Budiarti | Skor |
| :--- | :--- | :--- |
| **Product (Produk)** | Kualitas desain visual tinggi, bebas eror, dan fungsional. | **95/100** |
| **Service (Layanan)** | Sangat membantu mahasiswa lain; penjelasan jernih dan mudah dipahami. | **90/100** |
| **Knowledge (Pengetahuan)**| Akurasi konseptual sempurna; menunjukkan pemahaman mendalam. | **98/100** |
| **Value (Nilai)** | Ulasan sejawat sangat positif; beberapa mahasiswa langsung menggunakannya untuk belajar. | **88/100** |
| **Environmental (Lingkungan)** | Artefak digital, mudah didistribusikan, dan berkelanjutan. | **100/100** |

Dosen memberikan Budiarti "mata uang digital" sebagai penghargaan, yang meningkatkan statusnya di dalam ekosistem kelas @219. Lebih penting lagi, umpan balik dari rekan-rekannya membantunya menyempurnakan karyanya.

#### **Tahap 3: Transformasi dan Dampak Jangka Panjang**

Hasil dari proses ini jauh melampaui nilai ujian.

*   **Dampak Personal:** Budiarti tidak lagi melihat dirinya sebagai mahasiswa yang berjuang, tetapi sebagai seorang **kreator nilai** dan "pakar junior" di bidang Teorema Bayes @66. Identitas profesionalnya menguat, dan motivasi intrinsiknya meroket, sejalan dengan temuan penelitian (<<LATEX_0>> untuk identitas profesional, <<LATEX_1>> untuk motivasi) @17.

*   **Dampak Komunitas (Value Co-Creation):** Inilah bukti paling kuat dari keberhasilan VCC. Semester berikutnya, ketika angkatan baru mengambil mata kuliah yang sama, **infografis Budiarti menjadi salah satu sumber belajar utama yang direkomendasikan oleh sistem**. Karyanya menjadi bagian dari **aset pengetahuan komunal** yang terus berkembang @51, sebuah fenomena yang terbukti ketika **16% dari 427 artefak** yang dibuat selama penelitian digunakan kembali oleh angkatan selanjutnya @3 @43.

*   **Dampak Karier:** Artefak Budiarti, beserta riwayat kontribusi dan ulasan sejawatnya di GitHub, kini menjadi bagian dari **portofolio nilai yang terverifikasi** @20. Saat melamar magang, ia dapat menunjukkan bukti nyata kemampuannya dalam komunikasi teknis dan kolaborasi—sesuatu yang jauh lebih berharga daripada transkrip nilai.

Perjalanan Budiarti adalah mikrokosmos dari visi VALORIZE: mengubah ruang kelas menjadi **"teater kehidupan yang megah"** @8, di mana mahasiswa tidak hanya belajar untuk ujian, tetapi belajar untuk menjadi insinyur dan arsitek peradaban yang kompeten, beretika, dan inovatif.
