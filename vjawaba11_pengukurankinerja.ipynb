# Pengukuran Kinerja

Kinerja dalam kerangka kerja ini diukur secara kuantitatif dan multidimensional untuk memastikan objektivitas dan validitas. Metrik utama yang digunakan adalah:

* Hasil Belajar: Peningkatan skor ujian (diukur dengan t-test dan ANOVA).
* VCC Index: Ini adalah metrik inti untuk mengukur penciptaan nilai dari setiap artefak yang dihasilkan mahasiswa. Indeks ini adalah skor komposit (0-100) yang dihitung berdasarkan Rubrik PSKVE lima dimensi (Produk, Layanan, Pengetahuan, Nilai Guna, dan Dampak Etis/Sosial). Targetnya adalah 75% artefak yang dihasilkan mencapai VCC Index ≥75.
* Ukuran Efek (Cohen’s d): Untuk mengukur dampak intervensi terhadap hasil belajar, penelitian menargetkan effect size d ≥ 0,80 (dianggap sebagai dampak yang besar) saat membandingkan kelompok eksperimen dengan kelompok kontrol pada variabel seperti motivasi intrinsik, identitas profesional, dan transfer pembelajaran.
* System Usability Scale (SUS): kualiltas sistem CKM-SE diukur menggunakan kuesioner SUS dengan target skor ≥70, yang menunjukkan tingkat ketergunaan yang dapat diterima  
* Akurasi Profiling AI: Kinerja PUDAL Engine dalam memprofilkan mahasiswa diukur dengan target akurasi ≥85%  
* Tingkat Penggunaan Kembali Artefak: Kinerja sistem dalam menciptakan aset pengetahuan komunitas diukur dari persentase artefak yang dibuat oleh satu angkatan dan digunakan kembali oleh angkatan berikutnya. 
* Reliabilitas Antar-Penilai: Untuk memastikan konsistensi penilaian VCC, reliabilitas antar-penilai diukur menggunakan statistik kappa Cohen dengan target ≥0,75
