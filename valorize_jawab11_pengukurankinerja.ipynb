### Pengukuran Kinerja Holistik dalam Kerangka TISE-VALORIZE

Dalam filsafat **Triune-Intelligence Smart Engineering (TISE)**, kinerja sebuah sistem—termasuk sistem pembelajaran CKM-SE—harus diukur secara komprehensif. Pengukuran ini diperlukan karena dua alasan utama:

1.  **Akurasi Adaptasi (PUDAL Engine)**: Sistem yang cerdas (Smart Engineering) harus mampu **belajar** dari tindakan masa lalu (*Learning-Evaluating* dalam siklus PUDAL). Jika kinerja tidak terukur, model AI tidak dapat diperbarui, dan personalisasi adaptif akan gagal [ ] [ ] [ ].
2.  **Akuntabilitas Nilai (PSKVE Engine)**: Tujuannya adalah menciptakan nilai holistik. Nilai yang abstrak harus diubah menjadi **proses rekayasa yang terukur** untuk membuktikan bahwa artefak rekayasa selaras dengan nilai-nilai manusia dan konteks lingkungan [ ] [ ].

---

### Metrik Pengukuran Kinerja yang Berlapis

Kinerja dalam penelitian CKM-SE diukur melalui empat lapisan validasi (berdasarkan kerangka **PICOC-ASTF**) dan serangkaian metrik yang terstruktur:

#### 1. Pengukuran Kinerja Kognitif dan Afektif (Lapisan Aplikasi)

Ini adalah hasil pembelajaran utama yang diukur dalam studi quasi-eksperimental dengan target ukuran efek tinggi (Cohen’s $d \ge 0.80$) [ ] [ ] [ ] [ ].

| Metrik Kinerja | Cara Pengukuran | Target Keberhasilan | Bukti Empiris |
| :--- | :--- | :--- | :--- |
| **Motivasi Intrinsik** | Diukur menggunakan **Intrinsic Motivation Inventory (IMI)**, menilai minat/kesenangan, kompetensi yang dirasakan, dan otonomi [ ] [ ]. | Ukuran efek $d \ge 0.80$ (Intervensi > Kontrol) [ ] [ ] [ ]. | Mencapai $d=0.97$ [ ] [ ]. |
| **Identitas Profesional** | Diukur dengan skala yang menilai persepsi diri sebagai insinyur, rasa memiliki pada komunitas teknik, dan pertimbangan etis [ ] [ ]. | Ukuran efek $d \ge 0.80$ (Intervensi > Kontrol) [ ] [ ] [ ]. | Mencapai $d=0.84$ [ ] [ ]. |
| **Transfer Pembelajaran** | Diukur melalui **Novel Problem-Solving Test (NPST)**, yang memerlukan penerapan prinsip inti pada konteks yang belum diajarkan di kelas (transfer jauh) [ ] [ ] [ ]. | Ukuran efek $d \ge 0.85$ (Target terbesar) [ ] [ ] [ ]. | Mencapai $d=0.95$ [ ] [ ]. |
| **Kinerja Akademik** | Skor ujian tengah semester (UTS) dan nilai akhir [ ] [ ]. | Peningkatan skor yang signifikan pada kelompok dengan paparan tinggi. | Nilai rata-rata kelompok K3 (paparan intensif) secara signifikan lebih tinggi (34.04) dibandingkan K1 (22.14) ($F=13.34, p<0.001$) [ ] [ ]. |

#### 2. Pengukuran Penciptaan Nilai Bersama (VCC)

Kinerja holistik diukur menggunakan **PSKVE Engine** dan **VCC Index**, yang menilai hasil kerja mahasiswa (artefak) berdasarkan lima dimensi nilai yang menyerupai dunia profesional [ ] [ ]:

| Dimensi Nilai (PSKVE) | Metrik Terukur | Contoh Artefak |
| :--- | :--- | :--- |
| **Product (P)** (Kualitas Artefak) | Ketepatan teknis, kejelasan, profesionalisme (0-20 poin) [ ] [ ]. | Apakah perhitungan statistik benar? Apakah presentasi rapi? [ ] [ ]. |
| **Service (S)** (Utilitas Layanan) | Kegunaan artefak bagi pengguna lain atau komunitas [ ] [ ]. | Apakah laporan dapat ditindaklanjuti oleh mitra industri (misalnya, UMKM)? [ ] [ ]. |
| **Knowledge (K)** (Penciptaan Pengetahuan) | Frekuensi berbagi pengetahuan, tingkat adaptasi oleh rekan sejawat [ ] [ ]. | Apakah metodologi diadopsi oleh mahasiswa lain untuk analisis mereka? [ ]. |
| **Value (V)** (Dampak Ekonomi/Sosial) | Jumlah *download* oleh rekan sejawat, dampak sosial yang diukur [ ] [ ]. | 12 *peer download*, 2 adaptasi di fakultas lain [ ] [ ]. |
| **Environmental (E)** (Keberlanjutan/Etika) | Perlindungan data, tidak ada bias gender, kepatuhan etis [ ] [ ] [ ]. | Data dianonimkan, persetujuan diperoleh, dan sistem mendukung keberlanjutan [ ]. |

**Target Keberhasilan VCC**:
*   **Indeks VCC**: Minimal **75/100** untuk $75\%$ atau lebih artefak [ ] [ ].
*   **Tingkat Penggunaan Ulang**: Tingkat artefak yang digunakan kembali oleh kohort berikutnya harus mencapai minimal **$16\%$** (dibuktikan dengan platform *analytics*) [ ] [ ] [ ] [ ] [ ].

#### 3. Pengukuran Kinerja Sistem dan AI

Ini adalah pengukuran untuk Layer Sistem (S) dan Layer Teknologi (T) dalam kerangka PICOC-ASTF, yang menilai seberapa baik artefak CKM-SE berfungsi [ ] [ ]:

| Metrik Kinerja Sistem | Tujuan Pengukuran | Target Keberhasilan |
| :--- | :--- | :--- |
| **Usability Scale (SUS)** | Mengukur kemudahan penggunaan sistem dan kepuasan pengguna [ ] [ ] [ ]. | Skor SUS minimal $\ge 70$ [ ] [ ]. |
| **Akurasi Profiling AI** | Mengukur seberapa akurat PUDAL Engine mengidentifikasi gaya belajar (misalnya, FSLSM) dan celah pengetahuan mahasiswa [ ] [ ]. | Akurasi *profiling* minimal $\ge 85\%$ [ ]. |
| **Kualitas Interaksi Kolaboratif** | Metrik penggunaan sistem, frekuensi login, durasi sesi, dan interaksi dengan fitur kolaboratif [ ] [ ]. | Bukti interaksi yang kuat, misalnya rata-rata **9.2 ulasan sejawat** per artefak [ ] [ ]. |
