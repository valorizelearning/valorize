# Penjelasan Istilah: SE dan CKM

## Rekayasa Cerdas (*Smart Engineering* - SE)
Rekayasa Cerdas (*Smart Engineering* - SE): Bukan sekadar rekayasa alat, melainkan paradigma rekayasa sistem yang mengintegrasikan tiga kecerdasan (***Triune-Intelligence***) ke dalam artefak sistem dan seluruh siklus hidup rekayasa untuk menciptakan solusi yang adaptif, responsif, dan berkelanjutan. Berbeda dari rekayasa tradisional yang berfokus pada pemecahan masalah teknis murni (misalnya, membuat produk yang lebih efisien), Rekayasa Cerdas bertujuan merancang **"teater kehidupan"** atau ekosistem yang lebih luas tempat manusia dapat berkembang.

* Kecerdasan Manusia (Homocordium/*Natural Intelligence*): Merepresentasikan nilai-nilai, etika, kreativitas, dan tujuan yang mendefinisikan kemanusiaan,
* Kecerdasan Buatan (Homologos/*Artificial Intelligence*): Merepresentasikan subsistem teknis yang meniru fungsi kognitif seperti belajar dan bernalar, dan
* Kecerdasan Alam (Biophilos) / Kecerdasan Alamiah/Sosial (*Natural/Social Intelligence*): Merepresentasikan konteks sosio-ekologis tempat sistem beroperasi, termasuk struktur sosial, budaya, dan institusi

Dalam penelitian ini, SE diwujudkan dalam CKM-SE (*Collaborative Knowledge Management based on Smart Engineering*), sebuah *Smart Artefact* yang adaptif dan sadar nilai.   

## Manajemen Pengetahuan Kolaboratif (*Collaborative Knowledge Management* - CKM)
Manajemen Pengetahuan Kolaboratif (*Collaborative Knowledge Management* - CKM): Berbeda dengan manajemen pengetahuan tradisional yang statis (gudang file), **CKM adalah proses dinamis di mana pengetahuan dipertukarkan (*exchange*), dikombinasikan (*combine*), dan dikreasikan ulang (*co-create*) dalam tim yang heterogen untuk menghasilkan nilai baru**. Ini merupakan evolusi dari sistem manajemen pengetahuan (KMS) tradisional yang kini menekankan fasilitasi kolaborasi dan interaksi sosial.

Dalam penelitian ini, konsep tersebut diwujudkan dalam sebuah **platform digital** yang disebut CKMS-SE (Collaborative Knowledge Management System based on Smart Engineering). **CKMS-SE adalah sebuah *Smart Artefact* produk digital cerdas yang dirancang untuk memfasilitasi pembelajaran personal dan kolaboratif melalui tiga komponen utama**:

* Core *Engine*: Platform dasar untuk manajemen pengguna dan konten.
* PUDAL *Engine*: Mesin kognitif berbasis AI untuk personalisasi pembelajaran (*Personalized Learning*/PL).
* PSKVE *Engine*: Mesin manajemen nilai untuk mengukur dan mengoptimalkan VCC.

Pembelajaran Personal adalah pendekatan instruksional yang menyesuaikan konten, kecepatan, dan metode pembelajaran dengan kebutuhan, minat, dan kemampuan unik setiap mahasiswa. Teknologi AI dan analitik pembelajaran adalah pendorong utama implementasinya.

| 2. | Sayed et al. ### Data Penelitian Mengenai Manajemen Pengetahuan (*Knowledge Management*)

Manajemen Pengetahuan (KM) adalah proses menangkap, mendistribusikan, dan secara efektif menggunakan pengetahuan dalam sebuah organisasi. Di pendidikan tinggi, KM bertujuan untuk meningkatkan pengajaran, penelitian, dan kegiatan administratif.

| No. | Peneliti (Tahun) | Judul/Topik Utama | Temuan Kunci / Konsep Utama | Konteks / Metodologi | 
| :-- | :--- | :--- | :--- | :--- | 
| 1. | Fernandez-Nieto et al. (2024) | KM di Pendidikan Tinggi untuk *Communities of Practice* (CoP) Pendidik | KM sangat penting untuk meningkatkan pemecahan masalah, penelitian, dan pengajaran. Namun, implementasinya menghadapi tantangan budaya (keengganan berbagi), kurangnya insentif, dan keterbatasan teknologi. Diperlukan pendekatan holistik yang berpusat pada manusia. | Lokakarya *co-design* dengan tim pengajar di sebuah institusi pendidikan tinggi Australia. | 
| 2. | Abdullah et al. (2006) | Arsitektur Sistem Manajemen Pengetahuan (KMS) untuk Pembelajaran Organisasional | Mengusulkan arsitektur KMS tujuh lapis (antarmuka, akses, kolaboratif, aplikasi, transport, integrasi, repositori) untuk mengubah sumber daya pengetahuan menjadi modal intelektual. | Tinjauan konseptual dan pengembangan prototipe. | 
| 3. | Mohammed et al. (2021) | Model Kesuksesan Sistem Manajemen Pengetahuan (KMS) | Mengusulkan dan memvalidasi model kesuksesan KMS di PT. Faktor-faktor seperti kualitas pengetahuan, kualitas sistem, dan kualitas layanan secara signifikan memengaruhi penggunaan dan kepuasan pengguna. | Survei pada 378 pengguna KMS di universitas-universitas Irak menggunakan *Structural Equation Modeling* (SEM). | 
| 4. | Alzahrani & Alghamdi (2023) | Kerangka Kerja untuk Meningkatkan Penggunaan Teknologi eLearning | Mengusulkan kerangka kerja yang mengadopsi prinsip *Enterprise Architecture* (EA) untuk menyelaraskan tujuan strategis dengan implementasi teknologi. Ini adalah prinsip inti KM untuk memastikan teknologi mendukung tujuan organisasi. | Menggunakan *Design Science Research* (DSR). |
