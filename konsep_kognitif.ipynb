# Kognitif

Kognitif bukan hanya tentang "pintar" atau "hafal materi", tetapi mencakup berbagai fungsi mental seperti:

Atensi (Perhatian): Kemampuan untuk fokus pada satu hal sambil mengabaikan gangguan.

Memori (Ingatan): Bagaimana kita menyimpan, mempertahankan, dan memanggil kembali informasi di masa lalu.

Persepsi: Bagaimana kita menerjemahkan informasi yang ditangkap oleh indra (penglihatan, pendengaran, dll) menjadi sebuah makna.

Bahasa: Proses memahami kata-kata dan cara kita menyusun kalimat untuk berkomunikasi.

Pemecahan Masalah (Problem Solving): Kemampuan untuk menemukan jalan keluar dari situasi sulit atau kompleks.

Pengambilan Keputusan: Kemampuan memilih satu tindakan dari berbagai pilihan yang ada.

## Kognitif dalam konteks pendidikan

Ranah kognitif sering dikaitkan dengan Taksonomi Bloom, yaitu tingkatan cara berpikir manusia:

Mengingat: Menyebutkan kembali fakta.

Memahami: Menjelaskan ide atau konsep.

Menerapkan: Menggunakan informasi di situasi baru.

Menganalisis: Menghubungkan bagian-bagian informasi.

Mengevaluasi: Menilai atau memberi opini berdasarkan kriteria.

Mencipta (Creating): Menghasilkan karya atau ide baru (ini adalah tingkat tertinggi).

# Beban Kognitif

Otak manusia memiliki keterbatasan dalam memproses informasi pada satu waktu (disebut working memory). Cognitive Load Theory (CLT) menyatakan bahwa jika otak diberi terlalu banyak beban sekaligus, proses belajar akan berhenti.

Dalam konteks penelitian Anda, ada tiga jenis beban kognitif yang harus diseimbangkan:

Intrinsic Load (Beban Intrinsik): Kerumitan materi teknik itu sendiri (misalnya: rumus kalkulus kompleks atau logika pemrograman Robotika). Ini tidak bisa dihilangkan, tapi bisa dikelola.

Extraneous Load (Beban Ekstrinsik): "Gangguan" dari cara materi disajikan. Contoh: Antarmuka perangkat lunak Smart Engineering yang membingungkan atau instruksi lab yang tidak jelas. Ini adalah beban buruk yang harus dikurangi.

Germane Load (Beban Relevan): Upaya mental yang digunakan untuk benar-benar memahami dan membangun pengetahuan. Ini adalah beban baik yang harus didorong.

# Hubungannya dengan Value Co-Creation (Penelitian ini)
Mengapa manajemen beban kognitif penting untuk mencapai Value Co-Creation?

Kolaborasi Membutuhkan "Ruang" Kognitif: Agar mahasiswa bisa berkolaborasi (Collaborative Knowledge Management), mereka butuh sisa kapasitas mental. Jika mereka sudah "stuck" (kognitif penuh) hanya untuk memahami materi dasar, mereka tidak akan punya energi untuk berdiskusi, memberikan ide, atau melakukan co-creation.

Knowledge Sharing yang Efektif: Dalam model TISE-VALORIZE Anda, jika sistem manajemen pengetahuannya terlalu rumit, proses berbagi ilmu antar mahasiswa/dosen tidak akan terjadi secara alami.
