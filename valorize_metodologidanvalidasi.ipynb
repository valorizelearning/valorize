## **Metodologi yang Kaku: Membangun Inovasi dengan Bukti Empiris**

Keberhasilan Kerangka CKM-SE didasarkan pada metodologi riset yang ketat, menjembatani kesenjangan antara potensi teoretis dan dampak praktis pedagogis. Kami menggunakan dua kerangka kerja utama untuk memastikan validitas dan keketatan.

#### **1. Design Science Research Methodology (DSRM)**

CKM-SE dikembangkan sebagai sebuah *Smart Artefact* mengikuti siklus Design Science Research Methodology (DSRM) dari Peffers dkk. (2007) [@289 @72]. DSRM adalah filosofi penelitian yang berfokus pada penciptaan dan evaluasi artefak untuk memecahkan masalah praktis sambil memberikan kontribusi teoretis [@68 @79].

Proses pengembangan kami mencakup tiga fase Seminar Kemajuan (SK) yang terstruktur:
| Tahap SK Ke- | Tujuan Utama | Status/Target |
| :---: | :--- | :--- |
| **SK I** | Desain Arsitektur Konseptual (TISE, PUDAL, PSKVE) | Selesai (Agustus–Desember 2025) [@28 @2 @1] |
| **SK II** | Pengembangan Prototipe dan Uji Usabilitas | Direncanakan (Target SUS <<LATEX_0>>) [@28 @4 @65] |
| **SK III** | Validasi Empiris Skala Penuh (Quasi-Experiment) | Direncanakan (Target <<LATEX_1>>) [@28 @4 @71] |

#### **2. Kerangka Validasi PICOC-ASTF Berlapis**

Untuk mengatasi keterbatasan penelitian terdahulu yang menggunakan metrik evaluasi yang terbatas [@43], kami menerapkan kerangka **PICOC-ASTF** [@84 @2 @30]. Kerangka ini memungkinkan validasi berlapis dari ide yang paling fundamental hingga dampak aplikasi dunia nyata:

*   **Lapisan Fundamental (F):** Menilai koherensi dan validitas teori konseptual (misalnya, TISE dan VALORIZE) melalui penilaian ahli [@218 @293].
*   **Lapisan Teknologi (T):** Mengukur akurasi algoritma AI untuk personalisasi dan performa sistem secara teknis [@293].
*   **Lapisan Sistem (S):** Mengevaluasi kepuasan pengguna (Usability Score <<LATEX_2>>), kualitas kolaborasi, dan peningkatan hasil belajar [@23 @293].
*   **Lapisan Aplikasi (A):** Mengukur dampak pada Penciptaan Nilai Bersama (VCC Index), tingkat penggunaan kembali artefak, dan transfer pembelajaran. Target kami adalah *effect size* <<LATEX_3>> untuk hasil belajar utama [@19 @23 @293].

## **Anatomi Inovasi: Bagaimana CKM-SE Bekerja**

CKM-SE adalah orkestrasi Kecerdasan Manusia, Buatan, dan Alamiah. Sistem ini terdiri dari dua mesin yang terintegrasi ke dalam platform inti (Open LMS) [@339 @337]:

### **1. PUDAL Engine (The Cognitive Layer)**

PUDAL (Perceive, Understand, Decision-Making, Act-Response, Learning-Evaluating) adalah mesin personalisasi adaptif kami. Ini berfungsi sebagai mekanisme kognitif AI yang memodelkan cara berpikir pakar rekayasa, memastikan setiap mahasiswa belajar pada jalur yang sesuai dengan gaya belajarnya [@13 @57 @2]:

*   **Perceive & Understand:** Menganalisis data interaksi (misalnya, waktu yang dihabiskan pada diagram vs. teks) untuk memetakan profil belajar mahasiswa berdasarkan dimensi seperti **Sensing/Intuitive, Visual/Auditory, Active/Reflective,** dan **Sequential/Global** (FSLSM) [@60 @49].
*   **Decide & Act:** Jika seorang mahasiswa memiliki preferensi **Visual** dan **Reflektif**, PUDAL akan merekomendasikan: 1) visualisasi interaktif, 2) simulasi praktis, dan **kemudian** barulah pengenalan rumus abstrak. Tindakan ini disesuaikan secara *real-time* untuk mengatasi miskonsepsi secara diagnostik [@53 @62].

### **2. PSKVE Engine (The Value Co-Creation Layer)**

PSKVE (Product, Service, Knowledge, Value, Environmental) adalah kerangka kerja 5-dimensi yang mengubah VCC dari konsep abstrak menjadi metrik rekayasa yang terukur (VCC Index 0-100) [@2 @55]. PSKVE memastikan bahwa insinyur masa depan mengevaluasi solusi secara holistik, melampaui kinerja teknis murni:

| Dimensi PSKVE | Fokus Pengukuran | Contoh Penerapan dalam Proyek |
| :--- | :--- | :--- |
| **P (Product)** | Kinerja teknis dan fungsional artefak. | Akurasi perhitungan, kebenaran implementasi kode [@55]. |
| **S (Service)** | Kemudahan penggunaan dan rekomendasi yang dapat ditindaklanjuti. | Apakah solusi yang dihasilkan mudah dipahami oleh audiens awam? [@55]. |
| **K (Knowledge)** | Dokumentasi, generalisasi, dan potensi penggunaan kembali. | Apakah artefak dapat digunakan oleh angkatan berikutnya? [@54]. |
| **V (Value)** | Persepsi nilai komunitas dan komitmen etis pencipta. | Pengakuan rekan sebaya melalui *Knowledge Marketplace* [@58 @54]. |
| **E (Environmental)** | Pertimbangan keberlanjutan dan dampak siklus hidup produk. | Mitigasi dampak lingkungan, skor keberlanjutan jangka panjang [@70 @153]. |

### [Halaman: Transformasi Profesional]

## **Membentuk Arsitek: *T-Shaped* Professionalisme**

Kerangka TISE-VALORIZE didasarkan pada pergeseran taksonomi rekayasa dari klasifikasi berbasis disiplin (*insinyur listrik*, *insinyur sipil*) menjadi klasifikasi berdasarkan **Lapisan Kontribusi (ASTF)** [@211 @209].

#### **Simulasi Dunia Nyata: Mengadopsi Alur Kerja Profesional**

CKM-SE secara eksplisit mengintegrasikan praktik industri nyata yang dikenal sebagai *Authentic Audience and Professional Workflow* untuk menutup kesenjangan antara akademisi dan realitas profesional [@323]:

*   **Agile/Scrum Workflow:** Semester dibagi menjadi "sprint" berjangka waktu yang selaras dengan arsitektur ASTF [@319]. Peninjauan sebaya (rata-rata 9,2 tinjauan per artefak) berfungsi sebagai *Sprint Review* profesional, menumbuhkan akuntabilitas dan keandalan [@323 @321].
*   **Version Control (Git/GitHub):** Semua dokumen dan kode dikelola menggunakan Git/GitHub, menciptakan bukti kepenulisan yang tidak dapat disangkal dan mendemonstrasikan praktik kolaboratif yang terverifikasi, yang penting untuk mendapatkan *Reliability Bonds* (Nilai Perlindungan) dalam kerangka VALORIZE 2.0 [@320 @380].
*   **Keahlian *T-Shaped*:** Model pembelajaran ini memastikan setiap lulusan mengembangkan **keahlian berbentuk T**—pengetahuan mendalam di satu area teknis (spesialisasi teknologi) dan pemahaman yang luas tentang bagaimana keahlian itu terintegrasi ke dalam sistem sosio-teknis yang lebih besar (Lapisan Sistem dan Aplikasi) [@213 @222].

## **PICOC-ASTF: Kerangka Evaluasi Berlapis untuk Rekayasa Cerdas**

Proyek VALORIZE/CKM-SE tidak hanya divalidasi dengan membandingkan nilai akhir. Mengingat artefak ini adalah sebuah sistem cerdas, kami menggunakan kerangka evaluasi berlapis yang dikembangkan secara spesifik, yaitu **PICOC-ASTF** [ @3 ] [ @28 ].

Kerangka ini menggabungkan prinsip *Design Science Research Methodology* (DSRM) dengan validasi ilmiah berbasis bukti, memastikan setiap klaim efektivitas diukur dari filosofi paling dasar hingga dampak aplikasi dunia nyata [ @3 ] [ @152 ].

| Lapisan (ASTF) | Fokus Evaluasi | Pertanyaan yang Dijawab | Metrik Kunci |
| :--- | :--- | :--- | :--- |
| **F: Fundamental** | Koherensi Teoretis | Apakah filosofi **VALORIZE** (VOE-VCC-TISE) konsisten dan didukung oleh teori-teori seperti SDT dan CLT? | Validasi Ahli (skor rata-rata ≥ 4/5) [ @154 ] [ @3 ] |
| **T: Technology** | Kinerja Teknologi | Apakah mesin kognitif berbasis AI (PUDAL Engine) berfungsi dengan akurat (misalnya, akurasi *profiling* gaya belajar) dan seberapa efektif alat kognitif (Peta Pengetahuan) mereduksi beban kognitif? | Akurasi Algoritma, Kinerja Sistem, Pengurangan *Cognitive Load* [ @154 ] |
| **S: System** | Usabilitas & Penerimaan | Apakah sistem CKM-SE mudah digunakan oleh mahasiswa dan dosen? Apakah meningkatkan motivasi dan kualitas kolaborasi? | Skor **SUS** (target ≥ 70) [ @13 ] [ @10 ], Motivasi Intrinsik, Identitas Profesional [ @6 ] |
| **A: Application** | Dampak Dunia Nyata | Apakah sistem terbukti meningkatkan hasil belajar dan menciptakan nilai holistik (VCC) yang berkelanjutan? | Ukuran efek (**Cohen’s d ≥ 0.80**) untuk Transfer Pembelajaran, Tingkat Penggunaan Ulang Artefak (target ≥ 20%), VCC Index [ @25 ] [ @12 ] |

Validasi melalui empat lapisan ini (Fundamental, Technology, System, Application) memastikan bahwa CKM-SE adalah *Smart Artifact* yang kokoh, baik secara konseptual, teknis, maupun fungsional [ @3 ] [ @29 ].

### [Halaman: Peta Jalan & Visi Jangka Panjang]

## **Visi 2029–2030: Menjadi Arsitek Pendidikan Nasional**

Pengembangan VALORIZE terstruktur dalam sebuah Peta Jalan multi-tahun yang bertujuan untuk mencapai dampak nasional dan hilirisasi sebagai produk teknologi pendidikan (*EdTech*).

#### **Fase 1: Validasi Awal & Purwarupa (Tahun 1: 2025–2026)**
*   **Fokus:** Mengembangkan kerangka konseptual (SK I), merancang dan menguji purwarupa *Alpha* CKM-SE (SK II), dan memvalidasi efektivitas awal melalui uji coba terbatas (piloting) [ @126 ] [ @113 ].
*   **Status Kunci:** Purwarupa *Alpha* fungsional selesai (Maret 2026) dan publikasi dua artikel ilmiah Q2 [ @127 ] [ @120 ].

#### **Fase 2: Validasi Skala Luas & Iterasi (Tahun 2: 2026–2027)**
*   **Fokus:** Melaksanakan studi kuasi-eksperimental yang lebih besar dan longitudinal (*full-scale validation*) di berbagai mata kuliah dan program studi (target N=135) [ @2 ] [ @6 ] [ @129 ].
*   **Kegiatan:** Membandingkan hasil belajar antara kelas VALORIZE dan kelas kontrol selama satu tahun penuh untuk mengukur dampak jangka panjang terhadap retensi pengetahuan dan motivasi [ @128 ].

#### **Fase 3: Pengembangan Lanjutan & Diseminasi Nasional (Tahun 3–4: 2027–2029)**
*   **Fokus:** Mentransformasi purwarupa menjadi produk stabil (*production-grade*) dengan fitur AI dan analitik pembelajaran yang lebih canggih [ @131 ].
*   **Kegiatan:** Mengembangkan modul pelatihan (*training-of-trainers*) dan lokakarya untuk dosen di perguruan tinggi lain yang ingin mengadopsi VALORIZE. Diseminasi hasil riset di konferensi internasional tingkat atas [ @130 ] [ @131 ].

#### **Fase 4: Potensi Hilirisasi & Dampak EdTech (Tahun 5: 2029–2030)**
*   **Fokus:** Menjajaki potensi komersialisasi platform VALORIZE sebagai produk *EdTech* yang dapat diadopsi secara luas di Indonesia [ @132 ].
*   **Tujuan Akhir:** Bekerja sama dengan unit inovasi universitas (misalnya, LPIK ITB) untuk mengembangkan model bisnis berkelanjutan (misalnya, model *Software-as-a-Service* atau lisensi institusional), sejalan dengan Visi Indonesia Emas 2045 [ @132 ].

Visi ini mengukuhkan VALORIZE sebagai solusi yang dirancang tidak hanya untuk satu kelas atau institusi, tetapi sebagai cetak biru yang dapat direplikasi untuk masa depan pendidikan tinggi teknik di Indonesia [ @27 ].
