# Uji Coba dan Keterlibatan
Uji coba kerangka kerja CKMS-SE dirancang secara komprehensif dan bertahap melalui tiga fase utama yang melibatkan partisipan dari:

* Mahasiswa program sarjana di Sekolah Teknik Elektro dan Informatika (STEI) Institut Teknologi Bandung (ITB) terlibat sebagai pengguna platform, pembuat konten, dan penilai rekan sejawat (*peer-reviewer*).
* Dosen: Sebagai fasilitator dan validator ahli.
* Mata Kuliah Probabilitas dan Statistika.
* Sistem AI (PUDAL): Terlibat aktif sebagai "asisten cerdas" yang memetakan profil belajar mahasiswa.

## Tahapan Uji Coba
### Fase 1: *Pilot Study* (Bulan 1-6)

* Partisipan: 25 mahasiswa dan 1 dosen dari satu mata kuliah di satu program studi
* Tujuan: Menguji coba konsep awal dalam skala kecil dan mengumpulkan umpan balik
* Keterlibatan: Partisipasi bersifat sukarela dengan insentif berupa poin tambahan

### Fase 2: *Extended Trial* (Bulan 6-12)

* Partisipan: 50 mahasiswa dan 2 dosen dari dua mata kuliah di dua program studi  
* Tujuan: Menguji coba dalam skala lebih luas dan memvalidasi prototipe. Ini mencakup usability testing dengan target System Usability Scale (SUS) â‰¥70  
* Keterlibatan: Sistem diintegrasikan dengan mata kuliah wajib, dan elemen gamifikasi serta peer mentoring digunakan untuk meningkatkan keterlibatan mahasiswa 

### Fase 3: *Full Implementation* dan *Quasi-Experiment* (Bulan 12-24)

* Partisipan: 135 mahasiswa dari tiga mata kuliah (misalnya, Probabilitas dan Statistika) yang dibagi menjadi kelompok eksperimen (n=75) yang menggunakan CKMS-SE dan kelompok kontrol (n=60) yang menggunakan metode konvensional. Dosen yang terlibat juga akan mendapatkan dukungan melalui program pengembangan fakultas  
* Tujuan: Memvalidasi efektivitas kerangka kerja secara empiris dalam skala penuh untuk mengukur dampak pada motivasi, identitas profesional, dan VCC 
* Keterlibatan: Keterlibatan bersifat wajib sebagai bagian dari mata kuliah. Dosen didukung melalui lokakarya dan peluang kolaborasi riset untuk memastikan implementasi yang efektif

Saat ini, tidak semua mahasiswa atau dosen di ITB terlibat, melainkan partisipan yang dipilih secara spesifik dari mata kuliah dan program studi yang relevan untuk memastikan validitas hasil uji coba.
