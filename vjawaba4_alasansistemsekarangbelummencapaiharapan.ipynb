# Masalah Sistem Saat Ini

## Kegagalan Sistem Konvensional
Sistem pembelajaran saat ini (LMS konvensional) belum mencapai harapan / gagal karena terjebak dalam paradigma **"Transfer Pengetahuan"** satu arah dan **"Satu Ukuran untuk Semua"**.

* Masalah 1 (Isolasi): *Personalized Learning* (PL) sering membuat mahasiswa belajar sendiri-sendiri, kehilangan aspek sosial.

* Masalah 2 (Inefisiensi): *Collaborative Learning* (CL) sering gagal karena menyatukan mahasiswa dengan kecepatan belajar yang terlalu timpang tanpa mediasi.

* Masalah 3 (Nir-Makna): Fokus pada nilai ujian (*grade*) membuat mahasiswa tidak merasa perlu menciptakan dampak nyata (*Value*) bagi orang lain. Sistem sekarang belum memiliki "mesin" (*engine*) untuk mengukur dampak sosial dari tugas mahasiswa.

## Penelitian Terkait

Analisis terhadap 54 penelitian terkait menunjukkan adanya kesenjangan signifikan antara kondisi sistem pembelajaran saat ini (*current state*) dan kondisi yang diharapkan (*desired state*). Sistem yang ada saat ini gagal memenuhi harapan karena beberapa keterbatasan fundamental:

* Model Masih Konseptual dan Belum Teruji: Banyak kerangka kerja, terutama terkait VCC, masih bersifat teoretis dan belum diuji secara empiris dalam situasi nyata. Akibatnya, panduan implementasi yang detail sangat terbatas.
* Tidak Mendukung Konten Buatan Mahasiswa: Sistem pembelajaran konvensional umumnya berfokus pada penyampaian konten dari dosen ke mahasiswa (content delivery) dan belum secara spesifik dirancang untuk mendukung, mengelola, dan menilai konten yang dibuat oleh mahasiswa (student-generated content)
* Fokus pada Transfer Pengetahuan, Bukan Kolaborasi: Pendekatan yang ada lebih menekankan transfer pengetahuan daripada pengembangan keterampilan kolaboratif. Studi tentang pembelajaran kolaboratif online seringkali terbatas pada sampel kecil dan tidak dapat digeneralisasi  
* Evaluasi Interaksi Manusia-AI yang Kurang: Belum ada mekanisme evaluasi yang komprehensif untuk mengukur efektivitas interaksi antara mahasiswa dan AI, sehingga sulit untuk memastikan apakah penggunaan AI benar-benar mendukung tujuan pedagogis  
* Fragmentasi Integrasi: Upaya untuk mengintegrasikan Pembelajaran Personal (PL), Pembelajaran Kolaboratif (CL), AI, dan Pendidikan Berorientasi Nilai (VOE) masih terfragmentasi dan belum holistik


## Analisis Celah Penelitian dan Bukti Pendukungnya

### Celah Penelitian dan Bukti Pendukung

#### Model Masih Konseptual dan Belum Teruji
**Celah Penelitian 1:** Banyak kerangka kerja, terutama terkait *Value Co-Creation* (VCC), masih bersifat teoretis dan belum diuji secara empiris dalam situasi nyata. Akibatnya, panduan implementasi yang detail sangat terbatas.

**Bukti dari Referensi:**

*   **Kerangka Kerja yang Belum Tervalidasi Secara Statistik:** Penelitian oleh Redmond et al. (2018) yang mengembangkan *Online Engagement Framework*, secara eksplisit menyatakan bahwa langkah selanjutnya adalah validasi. Mereka menulis, "Pertama, kerangka kerja ini belum divalidasi secara statistik" dan mengundang peneliti lain untuk menggunakan kerangka kerja tersebut guna "membantu validasi". Ini menunjukkan model tersebut diajukan sebagai konsep awal.
*   **Model yang Membutuhkan Pengujian Eksperimental:** Studi Robayo-Pinzon et al. (2024) mengenai peran AI dalam VCC merekomendasikan agar "penelitian di masa depan mengambil pendekatan penelitian eksperimental atau tindakan melalui penerapan nyata di konteks pembelajaran universitas," yang mengindikasikan bahwa model mereka belum diuji secara praktis.
*   **Judul yang Menunjukkan Sifat Konseptual:** Penelitian oleh Dollinger, Lodge, & Coates (2018) berjudul *"Co-creation in higher education: towards a conceptual model"*. Judul itu sendiri menyiratkan bahwa tujuannya adalah membangun model konseptual, bukan menyajikan hasil pengujian empiris.
*   **Kesenjangan antara Teori dan Implementasi:** Wu et al. (2024) menyoroti bahwa model akademis harus "beralih dari konstruksi teoretis ke implementasi praktis", yang menunjukkan adanya kesenjangan antara teori dan aplikasi di dunia nyata.

#### Tidak Mendukung Konten Buatan Mahasiswa (*Student-Generated Content*)
**Celah Penelitian 2:** Sistem pembelajaran konvensional umumnya berfokus pada penyampaian konten dari dosen ke mahasiswa (*content delivery*) dan belum dirancang untuk mendukung, mengelola, dan menilai konten yang dibuat oleh mahasiswa.

**Bukti dari Referensi:**

*   **Pergeseran dari Konsumen ke Produsen:** Moro et al. (2024) menyatakan bahwa institusi pendidikan tinggi sedang "bermigrasi dari pendekatan didaktik... dan beralih ke desain pembelajaran di mana mahasiswa menjadi produsen, bukan konsumen konten." Fakta bahwa ini adalah sebuah "migrasi" menyiratkan bahwa keadaan umum saat ini adalah mahasiswa sebagai konsumen.
*   **Melampaui Transfer Pengetahuan:** Jamil et al. (2025) menekankan perlunya "melampaui transmisi pengetahuan" dan bergerak menuju "pertukaran pengetahuan yang dinamis, bukan transfer pengetahuan satu arah," yang menegaskan bahwa model dominan saat ini adalah transfer satu arah.

#### Fokus pada Transfer Pengetahuan, Bukan Kolaborasi
**Celah Penelitian 3:** Pendekatan yang ada lebih menekankan transfer pengetahuan daripada pengembangan keterampilan kolaboratif. Studi tentang pembelajaran kolaboratif daring seringkali terbatas pada sampel kecil dan tidak dapat digeneralisasi.

**Bukti dari Referensi:**

*   **Dominasi Transfer Pengetahuan:**
    > "Institusi tersier sedang bermigrasi dari pendekatan didaktik dan berpusat pada guru, dan sebaliknya, beralih ke desain pembelajaran yang mendorong pembelajaran langsung dan keterlibatan mahasiswa." (Moro et al., 2024). Pendekatan "didaktik" adalah ciri khas model transfer pengetahuan.
*   **Keterbatasan Studi Kolaboratif (Sampel Kecil):** Beberapa studi secara eksplisit mengakui keterbatasan ini.
    > "Keterbatasan pertama dari penelitian ini adalah ukuran sampel kelompok mahasiswa (4 kelompok menghasilkan 24 aktivitas kolaboratif)... Ukuran sampel yang kecil mungkin menjadi alasan... Oleh karena itu, penelitian empiris di masa depan perlu memperluas ukuran sampel." (Ouyang & Xu, 2023).
    > "beberapa temuan tidak boleh digeneralisasi ke konteks dan disiplin lain tanpa penelitian lebih lanjut. Penelitian lebih lanjut... diperlukan untuk memperluas bukti temuan... terutama yang memiliki jumlah mahasiswa lebih tinggi." (Tan et al., 2022).

#### Evaluasi Interaksi Manusia-AI yang Kurang
**Celah Penelitian 4:** Belum ada mekanisme evaluasi yang komprehensif untuk mengukur efektivitas interaksi antara mahasiswa dan AI, sehingga sulit memastikan apakah penggunaan AI benar-benar mendukung tujuan pedagogis.

**Bukti dari Referensi:**

*   **Fokus pada Akurasi Algoritma, Bukan Dampak Pedagogis:**
    > "Sebagian besar model prediksi AI yang ada fokus pada pengembangan dan optimalisasi akurasi algoritma AI daripada menerapkan model AI untuk memberikan umpan balik berkelanjutan kepada mahasiswa dan meningkatkan kualitas belajar mereka." (Ouyang et al., 2023).
*   **Kesenjangan Implementasi:** Zhang (2021) mengidentifikasi "kesenjangan kritis antara apa yang bisa dilakukan teknologi AIEd dan bagaimana teknologi tersebut sebenarnya diimplementasikan dalam pengaturan pendidikan yang otentik."
*   **Metrik Evaluasi yang Tidak Memadai:**
    > "Metrik penilaian konvensional gagal menangkap kemajuan belajar komprehensif mahasiswa dan efikasi sebenarnya dari sistem pembelajaran yang dipersonalisasi... [dan] secara tidak memadai mencerminkan dampak dunia nyata." (Wu et al., 2024).
*   **Kurangnya Refleksi Kritis:** Ullrich et al. (2022) menyimpulkan adanya "kurangnya refleksi kritis terhadap tantangan dan risiko masing-masing" dari penerapan AI dalam layanan dukungan akademik.

#### Fragmentasi Integrasi
**Celah Penelitian 5:** Upaya untuk mengintegrasikan Pembelajaran Personal (PL), Pembelajaran Kolaboratif (CL), AI, dan Pendidikan Berorientasi Nilai (VOE) masih terfragmentasi dan belum holistik.

**Bukti dari Referensi:**

*   **Implementasi yang Tersebar (*Scattered*):** Alzahrani & Alghamdi (2023) mengidentifikasi tantangan utama sebagai "implementasi dan penggunaan teknologi eLearning yang tersebar di antara para pemangku kepentingan," yang merupakan sinonim dari terfragmentasi.
*   **Kurangnya Solusi Komprehensif:** Deev & Finogeev (2023), dalam mengusulkan model pendidikan konvergen (terintegrasi), menyatakan bahwa institusi menghadapi "kurangnya panduan metodologis," "tidak ada mekanisme terpadu," dan "tidak ada solusi komprehensif" untuk implementasi.
*   **Pemisahan Proses Personalisasi dan Kolaborasi:** Abri et al. (2020) secara spesifik mengkritik sistem yang ada karena "memisahkan proses personalisasi dan kolaborasi" dan mengusulkan kerangka kerja untuk mengatasi kesenjangan ini.
*   **Fragmentasi Interdisipliner:** Wu et al. (2024) mencatat bahwa literatur saat ini menunjukkan "lebih sedikit referensi ke teori psikologi dan pendidikan, aplikasi terintegrasi data multimodal yang terbatas, dan eksplorasi faktor sosial yang relatif dangkal," yang menyoroti fragmentasi antara teknologi dan ilmu lainnya.
