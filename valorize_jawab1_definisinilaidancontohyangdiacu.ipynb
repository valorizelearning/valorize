## Definisi Nilai dalam Konteks Pembelajaran Rekayasa

Model pembelajaran VALORIZE/CKMS-SE secara fundamental berfokus pada pergeseran tujuan pendidikan dari penguasaan konten teknis (*knowing*) menuju **pembentukan identitas profesional** dan penciptaan nilai (*becoming*) [ @161 @210 @333 ].

### 1. Value-Oriented Education (VOE)
**VOE** adalah pendekatan pedagogis komprehensif yang melampaui pembelajaran akademik konvensional yang hanya berfokus pada transfer pengetahuan dan pencapaian kompetensi [ @122 @129 ].

*   **Definisi:** VOE adalah pendekatan pendidikan yang menempatkan penciptaan nilai (akademik, sosial, budaya, ekonomi, dan keberlanjutan) sebagai tujuan inti (*core objective*) dalam desain dan implementasi pembelajaran [ @33 @130 @132 ]. Sistem ini menekankan pentingnya nilai-nilai fundamental seperti empati, inklusivitas, dan keadilan dalam proses pembelajaran [ @122 @129 ].

### 2. Value Co-Creation (VCC)
**VCC** adalah mekanisme operasional untuk merealisasikan VOE [ @107 @129 ].

*   **Definisi:** VCC adalah proses kolaboratif antara mahasiswa, dosen, dan pemangku kepentingan (seperti industri) dalam menciptakan nilai pembelajaran yang saling menguntungkan, terukur, dan menghasilkan luaran yang memenuhi lima dimensi PSKVE [ @33 @34 ]. Kolaborasi ini menghasilkan output yang nilainya melebihi jumlah kontribusi individu [ @136 @150 ].

## Dimensi dan Contoh Nilai yang Diacu

Dalam kerangka kerja CKM-SE, penciptaan nilai holistik diukur dan dikelola melalui **PSKVE Engine** [ @135 @102 @292 ]. PSKVE mengukur nilai melalui lima dimensi energi yang saling terkait [ @292 @293 @280 ]:

| Dimensi | Fokus Pengukuran | Contoh Artefak Nilai dalam Probabilitas & Statistika |
| :--- | :--- | :--- |
| **P (Product)** | Kapasitas kerja fisik, menghasilkan artefak nyata, ketepatan teknis, dan kualitas [ @280 @293 @76 ]. | Toolkit Kontrol Proses Statistik (kode R, dokumentasi) yang bebas eror dan fungsional [ @79 @90 ]. |
| **S (Service)** | Kapasitas melayani dan memenuhi kebutuhan manusia hingga tercapai kepuasan; kegunaan praktis bagi orang lain [ @280 @293 @76 ]. | Penjelasan yang sangat jelas di forum, atau panduan langkah-demi-langkah (Peta Pengetahuan) yang dapat langsung diterapkan oleh rekan sejawat [ @151 @41 ]. |
| **K (Knowledge)** | Energi intelektual, kapasitas kreasi dan berbagi pengetahuan, dan kemampuan untuk digunakan kembali (*reusability*) [ @280 @293 @77 ]. | Metodologi analisis data yang dapat diadaptasi untuk konteks berbeda (misalnya, dari data kualitas air ke data risiko diabetes) [ @41 @81 @90 ]. |
| **V (Value)** | Kapasitas untuk merepresentasikan atau menukarkan nilai ekonomi, sosial, atau budaya; pengakuan komunitas [ @280 @293 @77 ]. | Pengakuan dari komunitas (misalnya, perencana kota) bahwa temuan analisis layak digunakan untuk pengambilan keputusan [ @90 @77 ]. |
| **E (Ethical/Environmental)** | Dampak dan interaksi dengan sistem alam; pertimbangan etika dan fokus keberlanjutan [ @280 @293 @78 ]. | Analisis data yang secara eksplisit membahas implikasi etika terhadap privasi data atau mempromosikan praktik berkelanjutan [ @14 @41 @92 ]. |

### Nilai Kemanusiaan Inti (Homocordium)

Sebagai pilar fundamental dari arsitektur **TISE (Triune-Intelligence Smart Engineering)**, nilai-nilai kemanusiaan inti (**Homocordium**) berfungsi sebagai kompas moral yang memandu seluruh proses rekayasa [ @281 @294 @296 ]. Nilai-nilai ini meliputi:

*   **Kemampuan Kognitif:** Berpikir kritis, pemecahan masalah, kreativitas, dan sintesis [ @47 ].
*   **Kecerdasan Emosional:** Empati, regulasi emosi, dan keterampilan interpersonal [ @47 ].
*   **Penalaran Etis:** Pengambilan keputusan berdasarkan nilai-nilai dan penilaian tentang benar dan salah [ @47 ].
*   **Motivasi Intrinsik:** Keinginan untuk menciptakan nilai, berkontribusi pada masyarakat, dan belajar untuk menguasai [ @47 ].
*   **Nilai Luhur:** Integritas dan etika profesional sebagai fondasi utama pembelajaran VALORIZE [ @137 @138 @212 ].

Penerapan nilai ini terlihat jelas dalam desain tugas, yang tidak hanya menanyakan "berapa confidence interval-nya?" tetapi juga "apa implikasi dari interval ini untuk *decision-making*?" dan "apakah *ethical* untuk menggunakan temuan statistik?" [ @14 ].
