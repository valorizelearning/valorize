# Domain Utama Penelitian
Penelitian ini berada dalam domain teknologi pendidikan, secara spesifik berfokus pada penerapan sistem cerdas untuk manajemen pengetahuan.
- Teknologi Pendidikan (*Educational Technology*): Sistem untuk pendidikan.
- Pendidikan Teknik (*Engineering Education*): Pendekatannya menggunakan *Smart Engineering* dan diterapkan di program studi teknik informatika.
- Sistem Informasi / Manajemen Pengetahuan (*Information Systems / Knowledge Management*): Penelitian tentang kerangkan kerja dan pengelolaan data/pengetahuan.

# Komponen Judul

- **Collaborative Knowledge Management** (Manajemen Pengetahuan) merujuk pada aspek sosial dalam berbagi ilmu.
- **Smart Engineering** (Teknologi Sistem Cerdas) merujuk pada Teknik Pengetahuan atau aplikasi AI yang sistematis untuk alur belajar yang lebih baik.
- **Value-Oriented Higher Education** (Pendidikan Nilai) menyoroti fokus pendidikan pada pembentukan karakter dan etika sebagai jantung dari sistem pendidikan.
