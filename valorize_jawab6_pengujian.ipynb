# Metodologi Uji Coba dan Keterlibatan Peserta

Uji coba kerangka CKM-SE tidak bersifat terbuka untuk semua orang, melainkan dirancang sebagai sebuah studi quasi-eksperimental yang terkontrol untuk mengukur dampak intervensi secara ilmiah.

---

#### 1. Desain dan Populasi Uji Coba

*   **Desain Penelitian**: Uji coba utama mengadopsi **desain quasi-eksperimental** dengan kelompok intervensi dan kelompok kontrol [ ] [ ]. Desain ini dipilih untuk membandingkan efektivitas CKM-SE dengan metode pembelajaran konvensional secara langsung [ ].

*   **Populasi Target**: Populasi terdiri dari **142 mahasiswa sarjana teknik tahun ketiga** di **Institut Teknologi Bandung** [ ]. Mereka terdaftar dalam dua mata kuliah inti yang berjalan paralel: **Probabilitas & Statistika** dan **Rekayasa Sistem** [ ].

*   **Pembagian Kelompok**: Mahasiswa dibagi menjadi dua kelompok yang seimbang:
    *   **Kelompok Intervensi (n=71)**: Mendapatkan akses penuh ke ekosistem CKM-SE [ ].
    *   **Kelompok Kontrol (n=71)**: Mengikuti pembelajaran dengan metode konvensional [ ].

*   **Durasi**: Intervensi ini dilaksanakan selama satu semester penuh, yaitu **12 minggu** [ ].

#### 2. Perlakuan (Intervensi) vs. Kontrol

Untuk memastikan perbandingan yang valid, kedua kelompok mendapatkan perlakuan yang sangat berbeda:

| Kelompok | Deskripsi Perlakuan |
| :--- | :--- |
| **Intervensi (CKM-SE)** | Mahasiswa terlibat dalam **"*Simulasi Profesi*"** berfidelitas tinggi [ ]. Mereka berperan sebagai "Insinyur Sistem TISE Junior" [ ], di mana mereka:<br>1. **Merespons Kebutuhan**: Membuat artefak pengetahuan melalui **Knowledge Marketplace** [ ].<br>2. **Menggunakan Alat Profesional**: Memvisualisasikan ide menggunakan **Knowledge Maps** berbasis AI dan mengelola proyek dengan **Agile/Scrum** serta **Git/GitHub** [ ] [ ].<br>3. **Mendapat Umpan Balik Adaptif**: Menerima bimbingan dari **Socratic AI Coaching** yang memandu, bukan memberi jawaban langsung [ ]. |
| **Kontrol (Konvensional)** | Mahasiswa mengikuti jalur pembelajaran linear (*one-size-fits-all*) tanpa personalisasi [ ]. Kolaborasi dilakukan melalui mekanisme standar seperti **forum diskusi di LMS** [ ] [ ], dan tugas dikembangkan secara individual tanpa kerangka VCC yang terstruktur [ ]. |

#### 3. Analisis Tambahan: Intensitas Penggunaan (Dose-Response)

Selain perbandingan langsung, sebuah analisis terpisah dilakukan pada dataset nilai Ujian Tengah Semester (UTS) Probabilitas dan Statistika (N=143) dari tiga kelas (K1, K2, K3) dengan tingkat paparan CKM-SE yang berbeda [ ].

*   **K1**: Paparan minimal terhadap fitur CKM-SE.
*   **K2**: Paparan tingkat menengah.
*   **K3**: Paparan paling intensif.

Hasilnya menunjukkan perbedaan kinerja yang sangat signifikan (**F = 13.34, p < 0.001**), di mana kelas dengan paparan tertinggi (K3, rata-rata nilai **34.04**) secara substansial mengungguli kelas dengan paparan terendah (K1, rata-rata nilai **22.14**) [ ]. Ini membuktikan bahwa **semakin tinggi intensitas keterlibatan dengan CKM-SE, semakin besar dampak positifnya terhadap hasil kognitif**.

---

#### 4. Keterbatasan Keterlibatan: Tidak Semua Peserta Meraih Hasil Maksimal

Penelitian ini secara jujur melaporkan bahwa kerangka kerja ini memiliki tantangan dan tidak semua peserta bisa langsung terlibat secara optimal.

*   **Kriteria Eksklusi**: Mahasiswa yang mengundurkan diri dari mata kuliah sebelum minggu ke-8 dikeluarkan dari analisis untuk memastikan data berasal dari peserta yang mendapatkan paparan intervensi yang memadai [ ].

*   **Tantangan Keterlibatan**: Ditemukan bahwa **sekitar 13% mahasiswa** dalam kelompok intervensi menunjukkan **peningkatan yang minimal** [ ]. Analisis lebih lanjut mengidentifikasi dua penyebab utama:
    1.  **Literasi Digital Awal yang Rendah**: Beberapa mahasiswa mengalami kesulitan dan kurva belajar yang curam saat harus beradaptasi dengan alat-alat standar industri seperti Git/GitHub [ ] [ ].
    2.  **Ketidaknyamanan dengan Evaluasi Sejawat**: Adanya rasa tidak nyaman atau penolakan terhadap proses tinjauan sejawat (*peer review*) menghambat partisipasi aktif mereka dalam siklus kolaborasi [ ].

*   **Solusi yang Diusulkan**: Berdasarkan temuan ini, penelitian merekomendasikan **solusi mitigasi** untuk implementasi di masa depan, yaitu:
    *   **Pelatihan Wajib**: Mengadakan lokakarya teknis khusus untuk alat-alat kolaboratif dan perangkat lunak industri ("Super Kendaraan") di awal semester [ ].
    *   **Perancah Berdiferensiasi (*Differentiated Scaffolding*)**: Menyediakan dukungan dan bimbingan yang disesuaikan dengan tingkat kesiapan teknis masing-masing mahasiswa [ ].

Dengan demikian, dapat disimpulkan bahwa uji coba dilakukan secara ketat dan terstruktur. Meskipun tidak semua peserta dapat langsung beradaptasi, penelitian ini berhasil mengidentifikasi akar masalahnya dan merumuskan strategi konkret untuk meningkatkan inklusivitas dan efektivitas di skala yang lebih luas.
