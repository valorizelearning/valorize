```markdown
# Judul: Penciptaan Nilai Bersama melalui Manajemen Pengetahuan Kolaboratif dalam Pendidikan Teknik Cerdas: Validasi Empiris Kerangka TISE-VALORIZE

**Penulis:** Meliana Christianti Johan†‡, Armein Z. R. Langi†, I Gusti Bagus Baskara Nugraha†, Radiant Victor Imbar‡ & Parlindungan Sinaga*
†Institut Teknologi Bandung, Indonesia
‡Maranatha Christian University, Indonesia
*Universitas Pendidikan Indonesia, Indonesia

## ABSTRAK

Kecerdasan buatan (AI) mengotomatisasi keterampilan teknik rutin namun gagal mengembangkan keterampilan manusia yang unik, yang merupakan krisis eksistensial dalam bidang teknik,. Paper ini memperkenalkan **Collaborative Knowledge Management System - Smart Engineering (CKMS-SE)** sebagai cetak biru untuk kerangka **Triune-Intelligence Smart Engineering (TISE)**, yang mengintegrasikan kecerdasan manusia, buatan, dan alami, dan **VALORIZE**, yang berfokus pada pendidikan berorientasi nilai. Melalui arsitektur mesin gandanya, CKMS-SE menggabungkan dua mesin untuk memungkinkan pembelajaran kolaboratif dan personalisasi serta menghasilkan penciptaan nilai bersama (*value co-creation*).

Studi **kuasi-eksperimental (N = 143)** menggunakan **Design Science Research Methodology (DSRM)**, kerangka kerja untuk memecahkan masalah kompleks dan merancang artefak inovatif, mengungkap dampak substansial. **Kelompok eksperimen skala besar (K2, N = 56) menunjukkan keuntungan 12,6% dalam kinerja kognitif** (*learning outcomes*) dibandingkan kelompok kontrol dasar (Rata-rata 24.93 vs 22.14). Hasil ini mengonfirmasi efektivitas intervensi dalam populasi dengan variabilitas tinggi (ANOVA **F = 13.34**; p < 0.001). Temuan ini menunjukkan kemampuan kerangka TISE-VALORIZE untuk menggeser fokus pendidikan dari penguasaan konten ke pengembangan profesional dalam kapabilitas humanistik esensial.

**Kata Kunci:** Kecerdasan buatan, sistem manajemen pengetahuan kolaboratif, pendidikan teknik, teknik cerdas, penciptaan nilai bersama.

## PENDAHULUAN

Pendidikan teknik menghadapi tantangan baru karena kemajuan AI menimbulkan pertanyaan tentang keterampilan lulusannya. Meskipun mahasiswa berhasil menguasai konten teknis, mereka sebagian besar tidak siap menghadapi tantangan teknik dalam merancang solusi dalam konteks sosial yang kompleks, membuat keputusan etis dalam ketidakpastian, dan berkolaborasi secara efektif lintas batas disiplin ilmu. AI mulai mengotomatisasi banyak keterampilan teknis yang menjadi fokus program teknik. Konvergensi ini menunjukkan kesenjangan pedagogis fundamental: pembelajaran tradisional berfokus pada pencapaian kompetensi dalam batasan materi, sumber daya, dan waktu yang terbatas tanpa mengartikulasikan dampak yang diharapkan dari kompetensi tersebut.

Penelitian ini mengusulkan platform yang mengintegrasikan Pembelajaran Kolaboratif dan Pembelajaran Personal, disinergikan oleh AI dalam konsep Pendidikan Berorientasi Nilai (*Value-Oriented Education/VOE*), untuk secara sistematis mempromosikan **Penciptaan Nilai Bersama (VCC)**,. VALORIZE menawarkan pendekatan baru dengan menggeser tujuan pendidikan dari sekadar menguasai konten menjadi **"membentuk sosok, karakter, dan pola pikir profesi"** (*becoming a figure*), menggunakan prinsip-prinsip **TISE Homocordium**.

## KERANGKA KONSEP DAN TEORI

### Inti Filosofis: Sintesis TISE-VALORIZE

Inti filosofis dari kerangka **TISE-VALORIZE** adalah pergeseran yang disengaja dari *knowing* (penguasaan konten) menjadi *becoming* (pengembangan karakter, pola pikir, dan penilaian profesional). Reorientasi ini secara strategis merespons perkembangan AI dan menegaskan bahwa nilai seorang insinyur berlabuh pada atribut manusia yang diwujudkan (**Homocordium**, yaitu etika, kreativitas, empati).

### Arsitektur Mesin Ganda CKMS-SE

CKMS-SE mengoperasionalisasikan TISE-VALORIZE melalui dua mesin khusus:

1.  **Mesin Kognitif (PUDAL Engine):** Mesin ini mengimplementasikan siklus peningkatan kognitif berkelanjutan. Siklus **PUDAL** (*Perceive, Understand, Decision-making and Planning, Act-Response, Learning-Evaluating*) bertindak sebagai "mesin mikro-evolusioner" yang memaksa peningkatan dan adaptasi berkelanjutan,,.
2.  **Mesin Ekonomi (PSKVE Engine):** Mesin ini mengelola dan mengoptimalkan Penciptaan Nilai Bersama (*Shared Value Creation/SVC*) dengan memastikan bahwa nilai yang diciptakan melampaui pembelajaran individu menuju pengembangan aset komunitas yang **terverifikasi**. Sistem ini mengoperasionalisasikan nilai komprehensif **PSKVE** (*Product, Service, Knowledge, Value, Environment*).

## METODE PENELITIAN

### A. Desain Studi dan Metodologi

1.  **Metodologi Pengembangan (DSRM):** Pengembangan dan evaluasi CKMS-SE mengikuti **Design Science Research Methodology (DSRM)** yang diartikulasikan oleh Peffers et al.. DSRM dipilih karena luaran penelitian ini adalah **artefak berbasis solusi**—sistem CKMS-SE,.
2.  **Desain Studi Empiris:** Studi ini mengadopsi **desain kuasi-eksperimental** yang dilakukan selama **satu semester (8 minggu)**.

### B. Populasi, Sampel, dan Kelompok Studi

1.  **Populasi:** Populasi total studi adalah **143 mahasiswa sarjana teknik** yang terdaftar dalam mata kuliah Probabilitas dan Statistik.
2.  **Pembagian Kelompok:**
    *   **Kelompok Intervensi/Eksperimen (K2):** N = 56 mahasiswa,.
    *   **Kelompok Kontrol (K1 dan K3):** N = 35 dan N = 52 mahasiswa.

## TEMUAN PENELITIAN

### A. Komponen Intervensi dan Skenario Eksperimen

Kelompok intervensi (K2) menggunakan metode **pembelajaran kolaboratif** yang distrukturkan sebagai **Simulasi Profesi** berdasarkan kerangka TISE. Siswa didorong untuk secara sukarela **membuat artefak** sebagai respons, yang kemudian diberi penghargaan menggunakan Portofolio Nilai PSKVE multi-aset.

#### i. Formalisasi Kognitif (Mesin PUDAL dan Multi-Graf)

CKMS-SE memodelkan proses kognitif ahli secara ketat melalui:
*   **Siklus PUDAL:** Seluruh proses kognitif distrukturkan oleh **siklus iteratif PUDAL**, yang memaksa peningkatan dan adaptasi berkelanjutan.
*   **Pemodelan Kognitif Formal:** Siswa diwajibkan menggunakan **Taksonomi Multi-Graf** formal, (misalnya, *Directed Acyclic Graphs/DAGs*, *Weighted Graphs*) untuk secara eksplisit memodelkan dan menilai proses kognitif multi-tahap insinyur ahli. Penggunaan *Weighted Graph* untuk membenarkan keputusan *trade-off* menjadikan proses penilaian ahli yang biasanya implisit menjadi **eksplisit dan dapat dipelajari**,.

#### ii. Alur Kerja Profesional dan Penciptaan Nilai (Mesin PSKVE)

1.  **Portofolio Nilai PSKVE Multi-Aset:** Portofolio Nilai PSKVE mensimulasikan **tiga pilar nilai teknik modern**: **Ekonomi** (diukur melalui *Productivity Units/PU*), **Protektif/Etika** (diukur melalui *Sustainability Bonds/SB*), dan **Sosial** (diukur melalui *Service Vouchers/SV*),,.
2.  **Alur Kerja Otentik:** Simulasi profesional dioperasionalisasikan dengan integrasi *sprint* **Agile/Scrum** dan penggunaan wajib **Git/GitHub**. Kewajiban ini menciptakan catatan kolaboratif yang autentik dan **terverifikasi** (*verifiable collaborative records*) praktik profesional,.

### B. Instrumen Pengukuran

Metrik dikalibrasi untuk secara eksplisit mengukur keberhasilan pergeseran filosofis dari *known* menjadi *being*.

1.  **Homodeus (Kognitif dan Transfer):** Diukur melalui penilaian **Transfer Learning**,.
2.  **Homocordium (Afektif dan Identitas):** Diukur menggunakan *Intrinsic Motivation Inventory* dan *Learning and Professional Identity in Practice Scale*.
3.  **Penciptaan Nilai Holistik (PSKVE):** Dinilai melalui metrik platform, termasuk tingkat pembuatan artefak (Manifestasi *Knowledge Credits/KC* dan *Productivity Units/PU*) dan frekuensi tinjauan sejawat (bukti *Service Vouchers/SV*),. Yang paling penting, **tingkat penggunaan kembali artefak oleh kohort berikutnya** diukur untuk membuktikan konversi pekerjaan akademik menjadi aset komunitas yang persisten dan terverifikasi yang menghasilkan *Economic Shares* (ES),,.

### C. Hasil Kuantitatif Kinerja Kognitif

Skor ujian tengah semester dari 143 mahasiswa dianalisis menggunakan uji **ANOVA satu arah**.

| Kelompok | Count | Average | Variance |
| :--- | :--- | :--- | :--- |
| K1 | 35 | 775 | 22.1428571 | 152.537815 |
| K2 (Eksperimen) | 56 | 1396 | 24.9285714 | 157.594805 |
| K3 | 52 | 1770 | 34.0384615 | 98.3906486 |

*Ringkasan Statistik ANOVA*

| Sumber Variasi | SS | df | MS | F | P-value |
| :--- | :--- | :--- | :--- | :--- | :--- |
| Between Groups | 3597.32168 | 2 | 1798.66084 | **13.3432357** | 4.96521E-06 |
| Within Groups | 18871.9231 | 140 | 134.799451 | | |

*   **Perbedaan Signifikan:** Hasil ANOVA menunjukkan perbedaan yang signifikan antar kelompok: **F = 13.34; p < 0.001**,,.
*   **Keunggulan K2:** Kelompok eksperimen skala besar (K2, N = 56) menunjukkan **keuntungan 12,6% dalam kinerja kognitif** (*learning outcomes*) dibandingkan dengan kelompok kontrol dasar (Rata-rata 24.93 vs 22.14),.
*   **Konfirmasi Intervensi:** Hasil ini secara empiris membuktikan efektivitas intervensi TISE-VALORIZE dalam populasi dengan variabilitas tinggi,. Peningkatan signifikan dalam Transfer Learning adalah hasil langsung dari formalisasi alur kerja kognitif melalui Taksonomi Multi-Graf.

## KESIMPULAN

Studi ini secara empiris memvalidasi kerangka CKMS-SE sebagai cetak biru operasional untuk filosofi **TISE-VALORIZE**. Temuan ini mengonfirmasi bahwa pergeseran fokus pendidikan teknik dari *knowing* menjadi *becoming*, melalui simulasi terstruktur, sangat efektif dalam menumbuhkan kompetensi manusia unik yang dibutuhkan di Era AI.

Peningkatan kinerja yang terverifikasi (**F=13.34**) menunjukkan bahwa kerangka ini memupuk **Homocordium** dengan memusatkan insentif pada Portofolio Nilai PSKVE, memformalkan pemikiran ahli melalui siklus iteratif **PUDAL** yang ketat dan **Taksonomi Multi-Graf**. Selain itu, kerangka ini menciptakan nilai yang dapat diverifikasi dengan mengimplementasikan alur kerja **Agile/Git**, yang mengubah tugas sementara menjadi aset intelektual yang berkontribusi pada komunitas.

Lulusan dari program TISE-VALORIZE muncul sebagai **"filsuf yang dapat dieksekusi"** (*executable philosophers*). Produk akhir mereka adalah **Portofolio Nilai yang Dapat Diverifikasi** yang kaya dan dapat diakses publik (biasanya di *GitHub Repository*). Portofolio ini memberikan bukti tak terbantahkan atas kompetensi mereka di semua lapisan TISE, menggabungkan peta kognitif yang terorganisir dan metrik nilai PSKVE yang tepat dengan riwayat Git yang transparan, menawarkan bukti nyata kolaborasi profesional mereka.
```
